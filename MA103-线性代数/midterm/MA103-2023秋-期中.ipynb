### 1-1 题目

Suppose that $\alpha_{1}, \alpha_{2}, \alpha_{3}$ are a basis for the nullspace of a matrix $A$, $N(A)$. Which of the following lists of vectors is also a basis of $N(A)$?

(A) $\alpha_{1}+\alpha_{2}-\alpha_{3}, \alpha_{1}+\alpha_{2}+5 \alpha_{3}, 4 \alpha_{1}+\alpha_{2}-2 \alpha_{3}$

(B) $\alpha_{1}+2 \alpha_{2}+\alpha_{3}, 2 \alpha_{1}+\alpha_{2}+2 \alpha_{3}, \alpha_{3}+\alpha_{1}+\alpha_{2}$

(C) $\alpha_{1}+\alpha_{2}, \alpha_{1}+\alpha_{2}+\alpha_{3}$

(D) $\alpha_{1}-\alpha_{2}, \alpha_{2}-\alpha_{3}, \alpha_{3}-\alpha_{1}$

```python
import sympy as sp
from IPython.display import display, Math

# 将每个选项的向量用基 {α₁,α₂,α₃} 的坐标表示，检查系数矩阵的秩
options = {
    'A': sp.Matrix([[1, 1, -1], [1, 1, 5], [4, 1, -2]]),
    'B': sp.Matrix([[1, 2, 1], [2, 1, 2], [1, 1, 1]]),
    'C': sp.Matrix([[1, 1, 0], [1, 1, 1]]),
    'D': sp.Matrix([[1, -1, 0], [0, 1, -1], [-1, 0, 1]]),
}

for label, M in options.items():
    r = M.rank()
    n_vecs = M.rows
    is_basis = "是基 (rank=3, 3个向量)" if r == n_vecs and n_vecs == 3 else ("只有2个向量，无法张成3维空间" if n_vecs < 3 else "线性相关")
    print(f"Option {label}: rank={r}, {is_basis}")

# 验证 D 中三向量之和为零
d = options['D']
print(f"\nD 中三向量之和: {d.row(0) + d.row(1) + d.row(2)}")
```

### 1-1 答案

**(A)**

选项 A 的系数矩阵满秩（rank=3），三向量线性无关，构成 $N(A)$ 的基。

选项 B 的系数矩阵秩为 2，线性相关。

选项 C 只有 2 个向量，无法张成 3 维空间。

选项 D 中三个向量之和为零，线性相关。

### 1-2 题目

Which of the following statements is correct?

(A) If the columns of $A$ are linearly independent, then $A x=b$ has exactly one solution for every $b$

(B) Any $5 \times 7$ matrix has linearly dependent columns

(C) If the columns of a matrix $A$ are linearly dependent, so are the rows

(D) The column space and row space of a $10 \times 12$ matrix may have differed dimensions

```python
import sympy as sp
from IPython.display import display, Math

# (A) 反例：列线性无关但非方阵，某些 b 无解
A_a = sp.Matrix([[1, 0], [0, 1], [0, 0]])
b_a = sp.Matrix([0, 0, 1])
print("(A) 反例：A 列无关，但对 b=[0,0,1]ᵀ 无解")
display(Math(r'A = ' + sp.latex(A_a) + r',\ b = ' + sp.latex(b_a)))

# (C) 反例：列相关但行无关
C2 = sp.Matrix([[1, 0, 1], [0, 1, 1]])
print(f"\n(C) 反例：2×3 矩阵 rank={C2.rank()}，3列在R²中必然相关，但2行线性无关")
display(Math(r'C = ' + sp.latex(C2)))

# (D) 列秩 = 行秩
print("\n(D) 错误：列秩 = 行秩，维数必须相等")
print("\n(B) 正确：5×7 矩阵有 7 个列向量在 R⁵ 中，必然线性相关")
```

### 1-2 答案

**(B)**

(D) 错误。列秩 = 行秩。

### 1-3 题目

Let

$$
\alpha_{1}=\begin{bmatrix}1\\4\\1\end{bmatrix},\ \alpha_{2}=\begin{bmatrix}2\\1\\-5\end{bmatrix},\ \alpha_{3}=\begin{bmatrix}6\\2\\-16\end{bmatrix},\ \beta=\begin{bmatrix}2\\t\\3\end{bmatrix}
$$

$\beta$ can be written as a linear combination of $\alpha_{1}, \alpha_{2}, \alpha_{3}$ if $t=($ )

(A) 1  (B) 3  (C) 6  (D) 9

```python
import sympy as sp
from IPython.display import display, Math

t = sp.Symbol('t')
alpha1 = sp.Matrix([1, 4, 1])
alpha2 = sp.Matrix([2, 1, -5])
alpha3 = sp.Matrix([6, 2, -16])
beta = sp.Matrix([2, t, 3])

A = sp.Matrix.hstack(alpha1, alpha2, alpha3)
aug = sp.Matrix.hstack(A, beta)

# 用 echelon_form 而非 rref：rref 会把一致性条件 (9-t) 除以自身规范化为 1，
# 导致 sp.solve(sp.Eq(1, 0), t) 返回空列表 []
print("增广矩阵的行阶梯形：")
echelon = aug.echelon_form()
display(Math(sp.latex(echelon)))

# 从最后一行得到一致性条件（令其为 0 即有解）
cond_expr = echelon[2, 3]
sol_t = sp.solve(cond_expr, t)
display(Math(r't = ' + sp.latex(sol_t)))
```

### 1-3 答案

**(D) 9**

将增广矩阵化为行阶梯形，从最后一行得到 $t-9=0$，故 $t=9$。

### 1-4 题目

Which of the following statements is correct?

(A) Suppose that $E A=B$ and $E$ is an invertible matrix, then the column space of $A$ and the column space of $B$ are the same

(B) Let $A$ be $n \times n$ matrix with rank 1, then $A^{n}=c A$, where $n$ is a positive integer and $c$ is a real number

(C) Let $A, B$ be symmetric matrices, then $A B$ is symmetric

(D) If $A$ is of full row rank, then $A x=0$ has only the zero solution

```python
import sympy as sp
from IPython.display import display, Math

# (B) 验证：rank-1 矩阵 A = u vᵀ，有 A² = (vᵀu)A，进而 Aⁿ = (vᵀu)ⁿ⁻¹ A = c A
n = 4
u = sp.Matrix([1, 2, 3, 4])
v = sp.Matrix([1, 1, 1, 1])
A_rank1 = u * v.T
c = (v.T * u)[0, 0]  # vᵀu = tr(A)

print(f"tr(A) = {c}")
display(Math(r'A = ' + sp.latex(A_rank1)))
display(Math(r'A^4 = ' + sp.latex(A_rank1**4)))
display(Math(r'c^3 A = ' + sp.latex(c**3 * A_rank1)))
print(f"A⁴ = c³A ? {A_rank1**4 == c**3 * A_rank1}")

# (C) 反例：对称矩阵乘积不一定对称
A_sym = sp.Matrix([[1, 0], [0, 2]])
B_sym = sp.Matrix([[0, 1], [1, 0]])
AB = A_sym * B_sym
print(f"\n(C) 反例 AB 对称? {AB == AB.T}")
display(Math(r'A = ' + sp.latex(A_sym) + r',\ B = ' + sp.latex(B_sym) + r',\ AB = ' + sp.latex(AB)))

# (D) 反例：full row rank 但列数 > 行数时，Ax=0 有非零解
A_full_row = sp.Matrix([[1, 0, 0], [0, 1, 0]])  # 2x3, rank=2
null = A_full_row.nullspace()
print(f"\n(D) 反例：A 行满秩(2×3, rank=2)，nullspace 非平凡: {null}")
```

### 1-4 答案

**(B)**

(A) 错误。$EA=B$ 中 $E$ 可逆时，行空间相同，但列空间可能不同。

(B) 正确。rank-1 矩阵可写为 $A=uv^T$，则 $A^2=(v^Tu)A$，归纳得 $A^n=(v^Tu)^{n-1}A=cA$。

(C) 错误。对称矩阵乘积对称当且仅当它们可交换。

(D) 错误。行满秩时若列数>行数，$Ax=0$ 有非零解。

### 1-5 题目

Let $A$ and $B$ be two $n \times n$ matrices. If $A$ is a non-zero matrix and $A B=0$, then

(A) $B A=0$

(B) $B=0$

(C) $(A+B)(A-B)=A^{2}-B^{2}$

(D) rank $B<n$

```python
import sympy as sp
from IPython.display import display, Math

# 反例验证
A = sp.Matrix([[0, 1], [0, 0]])
B = sp.Matrix([[1, 0], [0, 0]])

print(f"AB = {(A*B).tolist()}")
print(f"BA = {(B*A).tolist()}")
print(f"\n(A) BA=0 ? {B*A == sp.zeros(2)}")
print(f"(B) B=0 ? {B == sp.zeros(2)}")

# (D) 论证
print(f"\nrank(A)={A.rank()}, dim N(A)={2-A.rank()} = n-rank(A)")
print(f"rank(B)={B.rank()} < n=2 ✓")
```

### 1-5 答案

**(D)**

(A) 错误。反例：$A=\begin{bmatrix}0&1\\0&0\end{bmatrix}$，$B=\begin{bmatrix}1&0\\0&0\end{bmatrix}$，$AB=0$ 但 $BA=\begin{bmatrix}0&1\\0&0\end{bmatrix}\neq0$。

(B) 错误。$B$ 可以非零（如上）。

(C) 错误。$(A+B)(A-B)=A^2-B^2$ 仅当 $AB=BA$。

(D) 正确。$AB=0$ 意味 $B$ 的每列属于 $N(A)$。$A\neq0$ 故 $\operatorname{rank}(A)\geq1$，$\dim N(A)=n-\operatorname{rank}(A)<n$，所以 $\operatorname{rank}(B)\leq\dim N(A)<n$。

Denote the vector space of $7 \times 7$ real matrices by $M_{7 \times 7}(R)$, and let $W$ be the subspace of $M_{7 \times 7}(R)$ consisting of skew-symmetric real matrices, then $\operatorname{dim} W=$ $\underline{\quad\quad\quad}$. A matrix $A$ is called skew symmetric if $A^{\top}=-A$.

### 2-2 题目

Let $A, B$ be two $n \times n$ invertible matrices. Suppose the inverse of $\begin{bmatrix}A & C \\ 0 & B\end{bmatrix}$ is $\begin{bmatrix}A^{-1} & D \\ 0 & B^{-1}\end{bmatrix}$, where $O$ is the $n \times n$ zero matrix. Then $D=\underline{\quad\quad\quad}$.

```python
import sympy as sp
from IPython.display import display, Math

# 用具体矩阵验证分块求逆公式
n = 2
A1 = sp.Matrix([[1, 2], [0, 1]])
B1 = sp.Matrix([[2, 0], [1, 1]])
C1 = sp.Matrix([[1, 1], [0, 1]])

# 构造分块矩阵并求逆
M = sp.BlockMatrix([[A1, C1], [sp.zeros(n), B1]]).as_explicit()
M_inv = M.inv()

# 提取右下角 2×2 块
D_block = M_inv[0:n, n:2*n]
D_formula = -A1.inv() @ C1 @ B1.inv()

print("分块矩阵 M = [A C; 0 B]")
display(Math(r'M = ' + sp.latex(M)))
display(Math(r'M^{-1} = ' + sp.latex(M_inv)))
display(Math(r'D = ' + sp.latex(D_block)))
display(Math(r'-A^{-1}CB^{-1} = ' + sp.latex(D_formula)))
print(f"D = -A⁻¹CB⁻¹ ? {D_block == D_formula}")
```

### 2-2 答案

**$-A^{-1} C B^{-1}$**

由分块矩阵求逆公式：

$$
\begin{bmatrix}A & C \\ 0 & B\end{bmatrix}^{-1} = \begin{bmatrix}A^{-1} & -A^{-1}CB^{-1} \\ 0 & B^{-1}\end{bmatrix}
$$

故 $D = -A^{-1} C B^{-1}$。

### 2-4 题目

Consider the system of linear equations:

$$
A x=b=\begin{cases}
x+2 y=1 \\
x-y=2 \\
y=-1
\end{cases}
$$

The least-squares solution for the system is $\underline{\quad\quad\quad}$.

```python
import sympy as sp
from IPython.display import display, Math

# 3个方程，2个未知数 x, y
# x + 2y = 1, x - y = 2, y = -1
A_mat = sp.Matrix([[1, 2], [1, -1], [0, 1]])
b = sp.Matrix([1, 2, -1])

# 正规方程 AᵀA x̂ = Aᵀb
ATA = A_mat.T @ A_mat
ATb = A_mat.T @ b

print("正规方程 AᵀA x̂ = Aᵀb:")
display(Math(r'A^T A = ' + sp.latex(ATA) + r',\quad A^T b = ' + sp.latex(ATb)))

x_ls = ATA.solve(ATb)
display(Math(r'\hat{x} = ' + sp.latex(x_ls)))
```

### 2-4 答案

$$
\begin{bmatrix}19/11 \\ -5/11\end{bmatrix}
$$

解法：$A = \begin{bmatrix}1&2\\1&-1\\0&1\end{bmatrix}$，$b = \begin{bmatrix}1\\2\\-1\end{bmatrix}$。解正规方程 $A^T A \hat{x} = A^T b$ 得 $\hat{x} = \begin{bmatrix}19/11 \\ -5/11\end{bmatrix}$。

H=\left\{

\right\}

A unit vector orthogonal to $H$ is $\underline{\quad\quad\quad}$.

## 四

Let

$$
A=\begin{bmatrix}
1 & -1 & -1 \\
2 & a & 1 \\
-1 & 1 & a
\end{bmatrix},\ B=
\begin{bmatrix}
2 & 2 \\
1 & a \\
-a-1 & -2
\end{bmatrix}
$$

For which value(s) of $a$, the matrix equation $A X=B$ has no solution, a unique solution, or infinitely many solutions? Where $X$ is a $3 \times 2$ matrix. Solve $A X=B$ if it has at least one solution.

```python
import sympy as sp
from IPython.display import display, Math

a = sp.Symbol('a')
A = sp.Matrix([[1, -1, -1], [2, a, 1], [-1, 1, a]])
B = sp.Matrix([[2, 2], [1, a], [-a-1, -2]])

# 增广矩阵 [A|B]
aug = sp.Matrix.hstack(A, B)

print("增广矩阵 [A|B] 及其行阶梯形：")
display(Math(sp.latex(aug)))

# 进行高斯消元
rref, pivots = aug.rref()
display(Math(r'\text{rref} = ' + sp.latex(rref)))

# 分析不同 a 的情况
print("\n=== 案例分析 ===")
print(f"det(A) 因子分解: {sp.factor(A.det())}")

# Case a = -2
print("\n--- a = -2 ---")
A_m2 = A.subs(a, -2)
B_m2 = B.subs(a, -2)
aug_m2 = sp.Matrix.hstack(A_m2, B_m2)
print(f"rank(A) = {A_m2.rank()}, rank([A|B]) = {aug_m2.rank()}")
print("rank(A) ≠ rank([A|B]) ⇒ 无解")

# Case a = 1
print("\n--- a = 1 ---")
A_1 = A.subs(a, 1)
B_1 = B.subs(a, 1)
aug_1 = sp.Matrix.hstack(A_1, B_1)
print(f"rank(A) = {A_1.rank()}, rank([A|B]) = {aug_1.rank()}")
rref_1, _ = aug_1.rref()
display(Math(r'\text{rref}(a=1) = ' + sp.latex(rref_1)))

# Case a ≠ -2, a ≠ 1
print("\n--- a ≠ -2 且 a ≠ 1 ---")
print("rank(A) = rank([A|B]) = 3 ⇒ 唯一解")

# 求解唯一解的表达式
x1 = sp.Matrix([1, 0, -1])  # 第一列的解
x2 = sp.Matrix([3*a/(a+2), (a-4)/(a+2), 0])  # 第二列的解
X_unique = sp.Matrix.hstack(x1, x2)
display(Math(r'X = ' + sp.latex(X_unique)))
```

### 四 答案

**case 1: $a=-2$** 无解。此时 $\operatorname{rank}(A)=2 \neq 3=\operatorname{rank}([A \mid B])$。

**case 2: $a \neq -2$ 且 $a \neq 1$** 有唯一解：

$$
X = \begin{bmatrix}
1 & \dfrac{3a}{a+2} \\[6pt]
0 & \dfrac{a-4}{a+2} \\[6pt]
-1 & 0
\end{bmatrix}
$$

**case 3: $a=1$** 有无穷多解：

$$
X = \begin{bmatrix}
1 & 1 \\
-k_1-1 & -k_2-1 \\
k_1 & k_2
\end{bmatrix},\quad k_1,k_2 \in \mathbb{R}
$$

## 五

Let $M_{2 \times 2}(\mathbb{R})$ be the vector space of $2 \times 2$ real matrices. Let

$$
A=\begin{bmatrix}1 & 0 \\ 0 & -1\end{bmatrix},\ 
B=\begin{bmatrix}0 & 1 \\ 0 & 0\end{bmatrix},\ 
C=\begin{bmatrix}0 & 0 \\ 1 & 0\end{bmatrix}
$$

Consider the map

$$
T: M_{2 \times 2}(R) \rightarrow R^{3},
T(X)=\begin{bmatrix}
\operatorname{tr}(A^{\top} X) \\
\operatorname{tr}(B^{\top} X) \\
\operatorname{tr}(C^{\top} X)
\end{bmatrix}
$$

for any $2 \times 2$ real matrix $X$, where $\operatorname{tr}(D)$ denotes the trace of $D$.

(A) Show that $T$ is a linear transformation

(B) Find the matrix representation of $T$ with respect to the ordered basis

$$
v_{1}=\begin{bmatrix}1 & 0 \\ 0 & 0\end{bmatrix},
v_{2}=\begin{bmatrix}0 & 0 \\ 0 & 1\end{bmatrix},
v_{3}=\begin{bmatrix}0 & 1 \\ 1 & 0\end{bmatrix},
v_{4}=\begin{bmatrix}0 & 1 \\ -1 & 0\end{bmatrix}
$$

for $M_{2 \times 2}(R)$ and the standard basis $e_1,e_2,e_3$ for $R^{3}$.

(C) Can we find a matrix $X$ such that $T(X)=\begin{bmatrix}1 \\ -2 \\ 1\end{bmatrix}$? If yes, please find one such matrix; otherwise, give an explanation.

```python
import sympy as sp
from IPython.display import display, Math

# 定义 A, B, C
A_mat = sp.Matrix([[1, 0], [0, -1]])
B_mat = sp.Matrix([[0, 1], [0, 0]])
C_mat = sp.Matrix([[0, 0], [1, 0]])

# 定义基
v1 = sp.Matrix([[1, 0], [0, 0]])
v2 = sp.Matrix([[0, 0], [0, 1]])
v3 = sp.Matrix([[0, 1], [1, 0]])
v4 = sp.Matrix([[0, 1], [-1, 0]])

# T(X) = [tr(AᵀX), tr(BᵀX), tr(CᵀX)]ᵀ
def T(X):
    return sp.Matrix([
        (A_mat.T @ X).trace(),
        (B_mat.T @ X).trace(),
        (C_mat.T @ X).trace()
    ])

# (B) 计算 T(v_i) 在标准基下的坐标
print("=== (B) 矩阵表示 ===")
Tv = []
for i, v in enumerate([v1, v2, v3, v4], 1):
    tv = T(v)
    Tv.append(tv)
    display(Math(r'T(v_' + str(i) + r') = ' + sp.latex(tv)))

M_rep = sp.Matrix.hstack(*Tv)
print("矩阵表示 [T]_{v,e}：")
display(Math(r'[T]_{v,e} = ' + sp.latex(M_rep)))

# (C) 解 T(X) = [1, -2, 1]ᵀ
print("\n=== (C) 求解 T(X) = [1, -2, 1]ᵀ ===")
target = sp.Matrix([1, -2, 1])

# X 在基 v 下的坐标为 c = [c1,c2,c3,c4]ᵀ，需求解 M_rep * c = target
aug_check = sp.Matrix.hstack(M_rep, target)
print(f"rank(M_rep) = {M_rep.rank()}, rank([M_rep|target]) = {aug_check.rank()}")

if M_rep.rank() == aug_check.rank():
    print("target 在 C(M_rep) 中，可解")
    c1, c2, c3, c4 = sp.symbols('c1 c2 c3 c4')
    sol_set = sp.linsolve((M_rep, target), (c1, c2, c3, c4))
    sol = next(iter(sol_set))
    display(Math(r'\text{通解坐标 } c = ' + sp.latex(sp.Matrix(sol))))

    # 取一个特解：令自由参数 c2 = 0
    c_vec = sp.Matrix([sol[0].subs(c2, 0), sol[1].subs(c2, 0), sol[2].subs(c2, 0), sol[3].subs(c2, 0)])
    display(Math(r'\text{取 } c_2 = 0,\ c = ' + sp.latex(c_vec)))

    X_sol = c_vec[0]*v1 + c_vec[1]*v2 + c_vec[2]*v3 + c_vec[3]*v4
    display(Math(r'X = ' + sp.latex(X_sol)))
    print(f"验证 T(X) = {T(X_sol).tolist()}")

    # 也可直接由 T(A), T(B), T(C) 构造一个更直观的特解
    print("\n另一种方法：注意到 T(A)=[2,0,0]ᵀ, T(B)=[0,1,0]ᵀ, T(C)=[0,0,1]ᵀ")
    X_alt = sp.Rational(1, 2) * A_mat - 2 * B_mat + C_mat
    display(Math(r'X = \frac{1}{2}A - 2B + C = ' + sp.latex(X_alt)))
    print(f"验证 T(X) = {T(X_alt).tolist()}")
else:
    print("target 不在 C(M_rep) 中，无解")
```

### 五 答案

**(A)** 对任意 $X,Y\in M_{2\times2}(\mathbb{R})$ 和 $c\in\mathbb{R}$：

$$
T(cX+Y) = \begin{bmatrix}
\operatorname{tr}(A^{\top}(cX+Y)) \\
\operatorname{tr}(B^{\top}(cX+Y)) \\
\operatorname{tr}(C^{\top}(cX+Y))
\end{bmatrix}
= c\begin{bmatrix}
\operatorname{tr}(A^{\top}X) \\
\operatorname{tr}(B^{\top}X) \\
\operatorname{tr}(C^{\top}X)
\end{bmatrix}
+ \begin{bmatrix}
\operatorname{tr}(A^{\top}Y) \\
\operatorname{tr}(B^{\top}Y) \\
\operatorname{tr}(C^{\top}Y)
\end{bmatrix}
= cT(X)+T(Y)
$$

**(B)** 矩阵表示为：

$$
[T]_{v,e} = \begin{bmatrix}
1 & -1 & 0 & 0 \\
0 & 0 & 1 & 1 \\
0 & 0 & 1 & -1
\end{bmatrix}
$$

**(C)** 可以。注意到 $T(A)=\begin{bmatrix}2\\0\\0\end{bmatrix}$，$T(B)=\begin{bmatrix}0\\1\\0\end{bmatrix}$，$T(C)=\begin{bmatrix}0\\0\\1\end{bmatrix}$，取

$$
X = \frac{1}{2}A - 2B + C = \begin{bmatrix}
\frac{1}{2} & -2 \\
1 & -\frac{1}{2}
\end{bmatrix}
$$

则有 $T(X) = \begin{bmatrix}1\\-2\\1\end{bmatrix}$。

## 六

Let $A$ be an $m \times n$ matrix, $B$ be an $m \times p$ matrix, and $C$ be an $q \times p$ matrix. Show that

$$
\operatorname{rank}\begin{bmatrix}
A & B \\
0 & C
\end{bmatrix} \geq \operatorname{rank} A + \operatorname{rank} C.
$$

where 0 is the $q \times n$ zero matrix.

```python
import sympy as sp
from IPython.display import display, Math
import numpy as np

# 用随机矩阵验证不等式
np.random.seed(42)
m, n, p, q = 4, 3, 2, 3

A_np = np.random.randint(-3, 4, (m, n))
B_np = np.random.randint(-3, 4, (m, p))
C_np = np.random.randint(-3, 4, (q, p))

A = sp.Matrix(A_np)
B = sp.Matrix(B_np)
C = sp.Matrix(C_np)

M = sp.Matrix([[A[i,j] if j < n else B[i,j-n] for j in range(n+p)] for i in range(m)] + \
             [[0 if j < n else C[i-m,j-n] for j in range(n+p)] for i in range(m, m+q)])

# 更好的构造方式
M_block = sp.BlockMatrix([[A, B], [sp.zeros(q, n), C]]).as_explicit()

rank_M = M_block.rank()
rank_A = A.rank()
rank_C = C.rank()

print(f"rank(A) = {rank_A}")
print(f"rank(C) = {rank_C}")
print(f"rank(M) = {rank_M}")
print(f"rank(M) ≥ rank(A) + rank(C) ? {rank_M >= rank_A + rank_C}")
display(Math(
    r'\operatorname{rank}\begin{bmatrix}A&B\\0&C\end{bmatrix}=' + str(rank_M) +
    r'\geq ' + str(rank_A) + '+' + str(rank_C) + '=' + str(rank_A+rank_C)
))
```

### 六 答案

**证明**：对 $A$ 和 $C$ 分别进行初等行、列变换化为标准形：

$$
D_1 = \begin{bmatrix}I_r & 0 \\ 0 & 0\end{bmatrix} \text{（对 $A$）},\quad
D_2 = \begin{bmatrix}I_s & 0 \\ 0 & 0\end{bmatrix} \text{（对 $C$）}
$$

其中 $r = \operatorname{rank} A$，$s = \operatorname{rank} C$。

对分块矩阵 $M = \begin{bmatrix}A & B \\ 0 & C\end{bmatrix}$ 实施相应的初等变换，可将 $M$ 化为：

$$
M_1 = \begin{bmatrix}D_1 & C_1 \\ 0 & D_2\end{bmatrix}
$$

再利用 $D_1$ 和 $D_2$ 的 pivot 消去 $C_1$ 中的非零元，得：

$$
M_2 = \begin{bmatrix}
I_r & 0 & 0 & 0 \\
0 & 0 & 0 & C_2 \\
0 & 0 & I_s & 0 \\
0 & 0 & 0 & 0
\end{bmatrix}
$$

因此

$$
\operatorname{rank} M = \operatorname{rank} M_2 = r + s + \operatorname{rank} C_2 \geq r + s = \operatorname{rank} A + \operatorname{rank} C.
$$

初等变换不改变秩，不等式成立。$\square$
