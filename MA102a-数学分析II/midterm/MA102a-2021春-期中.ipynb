### 1-2 题目

$\lim\limits_{x \rightarrow +\infty, y \rightarrow +\infty} \left(1+\frac{1}{x}\right)^{\frac{x}{x+y}}$

```python
import sympy as sp
x, y = sp.symbols('x y', positive=True)
expr = (1 + 1 / x) ** (x / (x + y))
result = sp.limit(sp.limit(expr, y, sp.oo), x, sp.oo)
sp.simplify(result)
```

### 1-2 答案

$\lim\limits_{\substack{x \rightarrow+\infty, y \rightarrow+\infty}}(1+\frac{1}{x})^{\frac{x}{x+y}}=e^{\lim \limits_{x \rightarrow+\infty,y \rightarrow+\infty} \frac{x}{x+y} \ln (1+\frac{1}{x})}=e^{\lim\limits_{x \rightarrow+\infty,y \rightarrow+\infty}\frac{1}{x+y}}=e^0=1$

### 1-3 题目

当 $(x, y) \neq (0,0)$ 时，$f(x, y)=x^2\ln(x^2+y^2)$，并且 $f(0,0)=0$，求 $\frac{\partial f}{\partial x}$ 和 $\frac{\partial f}{\partial y}$。

```python
import sympy as sp
from IPython.display import display, Math

x, y, t = sp.symbols('x y t', real=True)
f = x**2 * sp.log(x**2 + y**2)

# 计算偏导（非原点）
dfx_nz = sp.simplify(sp.diff(f, x))
dfy_nz = sp.simplify(sp.diff(f, y))

# 计算偏导（原点，按定义）
dfx0 = sp.limit((t**2 * sp.log(t**2)) / t, t, 0)
dfy0 = sp.Integer(0)

# 构造 Piecewise 对象
# 显式指定条件以确保 SymPy 会渲染为 cases 格式
cond = sp.Or(sp.Unequality(x, 0), sp.Unequality(y, 0))
dfx_pw = sp.Piecewise((dfx_nz, cond), (dfx0, True))
dfy_pw = sp.Piecewise((dfy_nz, cond), (dfy0, True))

# 定义函数符号用于美观显示
func_f = sp.Function('f')(x, y)
display(sp.Eq(sp.Derivative(func_f, x), dfx_pw))
display(sp.Eq(sp.Derivative(func_f, y), dfy_pw))
```

### 1-3 答案

$\frac{\partial f}{\partial x}(x, y)=2 x \ln (x^{2}+y^{2})+\frac{2 x^{3}}{x^{2}+y^{2}}((x, y) \neq(0,0))$

$\frac{\partial f}{\partial x}(0,0)=\lim\limits_{t \rightarrow 0} 2 t \ln|t|=0$

$\frac{\partial f}{\partial y}(x, y)=\frac{2 x^{2} y}{x^{2}+y^{2}} \quad((x, y) \neq(0,0))$

$\frac{\partial f}{\partial y}(0,0)=0$

$\frac{\partial f}{\partial x}(x, y)= \begin{cases}2 x \ln (x^{2}+y^{2})+\frac{2 x^{3}}{x^{2}+y^{2}}, & (x, y) \neq(0,0), \\ 0, & (x, y)=(0,0)\end{cases}$

$\frac{\partial f}{\partial y}(x, y)= \begin{cases}\frac{2 x^{2} y}{x^{2}+y^{2}}, & (x, y) \neq(0,0), \\ 0 & (x, y)=(0,0)\end{cases}$

### 2 题目

给出下列极限的定义：$\lim\limits_{x \rightarrow a, y \rightarrow +\infty} f(x, y)=A$，其中 $0<a<+\infty$；并用此定义证明

$$
\lim\limits_{x \rightarrow a, y \rightarrow +\infty} \frac{2x^5+3xy^2}{x+y^2}=3a.
$$

```python
import sympy as sp
x, y, a = sp.symbols('x y a', positive=True)
expr = (2 * x**5 + 3 * x * y**2) / (x + y**2)
sp.simplify(sp.limit(sp.limit(expr, y, sp.oo), x, a))
```

### 2 答案

$\forall \varepsilon>0, \exists \delta>0, M>0$, 使得当 $0<|x-a|<\delta, y>M$ 时, $|f(x, y)-A|<\varepsilon$

证明: $\forall \varepsilon>0, \exists \delta=\min (\frac{\varepsilon}{6}, a), M=2 a \sqrt{\frac{2 \max (|16 a^{3}-3|,3)}{\varepsilon}}$

$0<|x-a|<\delta, y>M$ 时, $|\frac{2 x^{5}+3 x y^{2}}{x+y^{2}}-3 a|=|3(x-a)+\frac{x^{2}(2 x^{3}-3)}{x+y^{2}}|$

$\leq 3|x-a|+\frac{x^{2}|2 x^{3}-3|}{x+y^{2}}<\frac{\varepsilon}{2}+\frac{x^{2}|2 x^{3}-3|}{y^{2}}<\frac{\varepsilon}{2}+\frac{4 a^{2} \max (|16 a^{3}-3|, 3)}{y^{2}}$

$<\frac{\varepsilon}{2}+\frac{\varepsilon}{2}=\varepsilon$

### 3 题目

求出曲线 $x=y^2-1$ 和 $x-y=11$ 在上半平面围成的图形绕 $x$ 轴旋转一周所得的旋转体体积。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 变量定义
x, y = sp.symbols('x y', real=True)

# 曲线定义：
# C1: x = y^2 - 1 (抛物线)
# C2: x - y = 11 => x = y + 11 (直线)
# 题目要求：上半平面 (y > 0)

# 1. 求交点 (确定 y 的积分范围)
# y^2 - 1 = y + 11 => y^2 - y - 12 = 0 => (y-4)(y+3) = 0
# y = 4 (上半平面) 或 y = -3 (舍去)
y_start = 0
y_end = 4

# 2. 计算体积 (绕 x 轴旋转)
# 使用 圆柱壳法 (Shell Method)
# V = 2 * pi * integrate( 半径 * 高度, (y, y_start, y_end) )
# 半径 = y
# 高度 = 右边界 - 左边界 = (y + 11) - (y^2 - 1)
radius = y
height = (y + 11) - (y**2 - 1)

volume = 2 * sp.pi * sp.integrate(radius * height, (y, y_start, y_end))
display(sp.Eq(sp.Symbol('V'), volume))

# 3. 绘图可视化
y_vals = np.linspace(0, 5, 400)
x1 = y_vals**2 - 1
x2 = y_vals + 11

plt.figure(figsize=(8, 4))
plt.plot(x1, y_vals, label='$x = y^2 - 1$')
plt.plot(x2, y_vals, label='$x = y + 11$')
plt.fill_betweenx(y_vals, x1, x2, where=(y_vals >= 0) & (y_vals <= 4), color='gray', alpha=0.3)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area in Upper Half-Plane')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### 3 答案

$V=\pi \int_{-1}^{15}(x+1) d x-\pi \int_{11}^{15}(x-11)^{2} d x=\frac{320}{3} \pi$

设 $f(x, y)=\frac{x^3}{x^2+y^2}$（当 $x^2+y^2>0$），并且 $f(0,0)=0$。

### 6-1 题目

证明 $f(x, y)$ 于 $(0,0)$ 处沿任何方向的方向导数都存在。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x, y = sp.symbols('x y', real=True)
f_expr = x**3 / (x**2 + y**2)

# 计算方向导数 (u = (cos(theta), sin(theta)))
t, theta = sp.symbols('t theta', real=True)
expr_pol = (t * sp.cos(theta)) ** 3 / ((t * sp.cos(theta)) ** 2 + (t * sp.sin(theta)) ** 2)
directional = sp.simplify(sp.limit(expr_pol / t, t, 0))
display(sp.Eq(sp.Symbol(r'\frac{\partial f}{\partial u}(0,0)'), directional))

# 3D 绘图可视化函数表面
f_func = sp.lambdify((x, y), f_expr, 'numpy')

X = np.linspace(-1, 1, 100)
Y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(X, Y)
# 避免原点除零，原点定义为0
Z = np.zeros_like(X)
mask = (X**2 + Y**2 > 1e-9)
Z[mask] = f_func(X[mask], Y[mask])

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)
ax.set_title('$f(x, y) = \\frac{x^3}{x^2+y^2}$')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```

### 6-1 答案

证明: $u=(\cos \theta, \sin \theta), \theta \in[0,2 \pi)$ 为所有方向,

$$
\frac{\partial f}{\partial n}(0,0)=\lim\limits_{t \rightarrow 0} \frac{f(t n)-f(0,0)}{t}=\cos^{3} \theta
$$

### 6-2 题目

判断 $f(x, y)$ 于 $(0,0)$ 处是否可微，并证明你的结论。

```python
import sympy as sp
x = sp.symbols('x', positive=True)
g_y0 = sp.simplify((x**3 / x**2 - x) / sp.sqrt(x**2))
g_yx = sp.simplify(((x**3 / (x**2 + x**2)) - x) / sp.sqrt(x**2 + x**2))
sp.Matrix([sp.limit(g_y0, x, 0), sp.limit(g_yx, x, 0)])
```

### 6-2 答案 

不可微. 证明如下:

$\lim\limits_{(x, y) \rightarrow(0,0)} \frac{f(x, y)-f(0,0)-\frac{\partial f}{\partial x}(0,0) x-\frac{\partial f}{\partial y}(0,0) y}{\sqrt{x^{2}+y^{2}}}=\lim\limits_{(x, y) \rightarrow(0,0)} \frac{\frac{x^{3}}{x^{2}+y^{2}}-x}{\sqrt{x^{2}+y^{2}}}$

$=\lim\limits_{(x, y) \rightarrow(0,0)}-\frac{x y^{2}}{(x^{2}+y^{2})^{\frac{3}{2}}}$

设 $g(x, y)=-\frac{x y^{2}}{(x^{2}+y^{2})^{\frac{3}{2}}}$

取 $s_{i}=(\frac{1}{i}, 0), t_{i}=(\frac{1}{i}, \frac{1}{i})$, 则 $s_{i} \rightarrow 0(i \rightarrow \infty), t_{i} \rightarrow 0(i \rightarrow \infty), g(s_{i})=0, g(t_{i})=-\frac{\sqrt{2}}{4}$

$\Rightarrow \lim\limits_{(x, y) \rightarrow(0,0)} g(x, y)$ 不存在

$\Rightarrow f(x, y)$ 于 $(0,0)$ 处不可微

### 7 题目

设 $f(x, y)=\sin(xy)$，$(x, y) \in \mathbb{R}^2$。

1. 证明：$f(x, y)$ 于区域 $D_a=\{(x, y) ; x^2+y^2<a\}$（$0<a<+\infty$）上一致连续；
2. 判断 $f(x, y)$ 是否于 $\mathbb{R}^2$ 上一致连续，并证明。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x, y = sp.symbols('x y', real=True)
f_val = sp.sin(x * y)

# 验证非一致连续的反例
n = sp.symbols('n', positive=True, integer=True)
p1 = (n, sp.pi/(2*n))
p2 = (n, sp.pi/(6*n))
dist = sp.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
diff = sp.Abs(f_val.subs({x: p1[0], y: p1[1]}) - f_val.subs({x: p2[0], y: y})) # 修正变量
diff_val = sp.Abs(sp.sin(p1[0]*p1[1]) - sp.sin(p2[0]*p2[1]))

display(Math(rf"\text{{Distance }} \delta_n = {sp.latex(sp.simplify(dist))} \to 0"))
display(Math(rf"\text{{Function Difference }} |f(P_1)-f(P_2)| = {sp.latex(sp.simplify(diff_val))}"))

# 绘制函数图像
X = np.linspace(-5, 5, 200)
Y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X * Y)

plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, Z, levels=20, cmap='RdBu')
plt.colorbar(cp)
# 使用原始字符串 (r'') 解决 \s 转义警告
plt.title(r'$f(x, y) = \sin(xy)$ Contour Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 7 答案

#### 7.1

证明:  $\bar{D}=\{(x, y) \mid x^{2}+y^{2} \leq a\}$ 是有界闭集, 因此是紧致集

又由于f是 $\bar{D}$ 上的连续函数, 故f在 $\bar{D}$ 上一致连续, 则 $f$ 在 $D$ 上也一致连续

#### 7.2

不一致连续. 证明如下:

取 $S_{i}=(i, \frac{\pi}{2 i}), t_{i}=(i, \frac{\pi}{6 i})$。则 $\|S_{i}-t_{i}\|=\frac{\pi}{3 i} \rightarrow 0(i \rightarrow \infty), f(S_{i})=1, f(t_{i})=\frac{1}{2}$,  $|f(S_{i})-f(t_{i})|=\frac{1}{2}$

$\Rightarrow f$ 在 $\mathbb{R}^{2}$ 上不一致连续


## 八

设 $f(x_1,x_2,\cdots,x_n)$ 是 $n$ 维 Euclid 空间中球面

$$
S=\{(x_1,x_2,\cdots,x_n); x_1^2+x_2^2+\cdots+x_n^2=a^2\}, \quad 0<a<+\infty
$$

上的连续函数。

### 8-1 题目

证明：存在 $S$ 中两点使得 $f$ 分别取得最大值 $M$ 与最小值 $m$。

### 8-1 答案 


$S \subseteq B_{2 a}(0)=\{x \in \mathbb{R}^{n} \mid\|x\|<2 a\}$

$\Rightarrow S$ 是有界集

$S^{c}=\{x \in \mathbb{R}^{n} \mid\|x\|<a 或 \|x\|>a\}$

对任意 $x \in S^{C}$,

若 $\|x\|<a$, 令 $r=a-\|x\|$, 对任意 $y \in B_{r}(x)$,

$\|y\| \leq\|y-x\|+\|x\|<a-\|x\|+\|x\|=a$

$\Rightarrow B_{r}(x) \subseteq S^{c}$

若 $\|x\|>a$, 令 $r=\|x\|-a$, 对任意 $y \in B_{r}(x)$

$\|y\| \geq\|x\|-\|x-y\|>\|x\|-(\|x\|-a)=a$

$\Rightarrow B_{r}(x) \subseteq S^{C}$

$\Rightarrow S^{c}$ 是开集

$\Rightarrow$ S是闭集

又由于S是有界集, 故S是紧致集

又由于f是连续的, 故 $f$ 在S上可以取到最大值和最小值

### 8-2 题目

当 $n=2$ 且 $M>m$ 时，证明：对任意实数 $\eta \in (m, M)$，存在至少两个球面上的互异点使得 $f=\eta$。

### 8-2 答案


设 $f(x)=M, f(y)=m, S=\{(x, y) \mid x=a \cos \theta, y=a \sin \theta\}$,  $x=(a \cos \theta_{1}, a \sin \theta_{1}), y=(a \cos \theta_{2}, a \sin \theta_{2})$, 不妨设 $\theta_{1}>\theta_{2}$ 且 $|\theta_{1}-\theta_{2}| < 2 \pi$

令 $S_{1}=\{(x, y) \mid x=a \cos \theta, y=a \sin \theta, \theta \in[\theta_{2}, \theta_{1}]\}$,

$S_{2}=\{(x, y) \mid x=a \cos \theta, y=a \sin \theta, \theta \in[\theta_{1}, \theta_{2}+2 \pi]\},$

对 $\forall x, y \in S_{1}, ~$ 设 $x=(a \cos \varphi_{1}, a \sin \varphi_{1}), y=(a \cos \varphi_{2}, a \sin \varphi_{2})$, 不妨设 $\varphi_{2}>\varphi_{1}$

则存在连续映射 $x=\varphi(t)=(a \cos t, a \sin t), t \in[\varphi_1, \varphi_2]$ 并且 $\varphi(\varphi_1)=x, \varphi(\varphi_2)=y$

$\Rightarrow S$ 是连通集

考虑$S_{1}$上的连续函数 $f$, 则 $\exists \xi_{1} \in S_{1}$, 使得 $f(\xi_{1})=\eta \in(f(y), f(x))=(m, M)$

同理,  $\exists \xi_{2} \in S_{2}$ 使得 $f(\xi_{2})=\eta \in(m, M)$

$\Rightarrow S$ 中至少存在两个互异的点使得$f$在这两点处函数值等于$\eta$

$\operatorname{QED}$
