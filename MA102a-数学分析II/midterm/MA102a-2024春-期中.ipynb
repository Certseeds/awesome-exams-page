### 1-1 题目

设 $z=e^{u} \ln v, u=\frac{x}{y}, v=x+4y$，求 $\frac{\partial z}{\partial x}$ 和 $\frac{\partial z}{\partial y}$。

```python
import sympy as sp
x, y = sp.symbols('x y', real=True, nonzero=True)
u = x / y
v = x + 4 * y
z = sp.exp(u) * sp.log(v)
dz_dx = sp.simplify(sp.diff(z, x))
dz_dy = sp.simplify(sp.diff(z, y))
sp.Matrix([dz_dx, dz_dy])
```

### 1-1 答案

$\frac{\partial z}{\partial x}=e^{\frac{x}{y}}\left(\frac{1}{y}\ln(x+4y)+\frac{1}{x+4y}\right)$

$\frac{\partial z}{\partial y}=e^{\frac{x}{y}}\left(-\frac{x}{y^2}\ln(x+4y)+\frac{4}{x+4y}\right)$

### 1-2 题目

求函数 $f(x,y)=\sqrt{1-x^2}+\sqrt{y^2-1}$ 于点 $D=(\frac{1}{2},-\frac{3}{2})$ 处，沿方向 $u=(\frac{\sqrt{3}}{2}, \frac{1}{2})$ 的方向导数 $\frac{\partial f}{\partial u}(D)$。

```python
import sympy as sp
x, y = sp.symbols('x y', real=True)
f = sp.sqrt(1 - x**2) + sp.sqrt(y**2 - 1)
grad_f = sp.Matrix([sp.diff(f, x), sp.diff(f, y)])
point = {x: sp.Rational(1, 2), y: -sp.Rational(3, 2)}
direction = sp.Matrix([sp.sqrt(3) / 2, sp.Rational(1, 2)])
sp.simplify(grad_f.subs(point).dot(direction))
```

### 1-2 答案

$\frac{\partial f}{\partial x}=-\frac{x}{\sqrt{1-x^2}}, \quad \frac{\partial f}{\partial y}=\frac{y}{\sqrt{y^2-1}}$

$f$ 在点 $D$ 的邻域内可微，因此方向导数等于梯度与方向向量的点积。

$\frac{\partial f}{\partial u}(D)=\left(\frac{\sqrt{3}}{2}, \frac{1}{2}\right) \cdot \left(-\frac{\sqrt{3}}{3}, -\frac{3\sqrt{5}}{5}\right)=-\frac{1}{2}-\frac{3\sqrt{5}}{10}$

### 1-4 题目

计算曲线 $x=a \cos t, y=b \sin t, z=0$ 的曲率，其中 $0<b<a, 0 \le t \le 2\pi$，并指出曲率的最大值与最小值。

```python
import sympy as sp
t, a, b = sp.symbols('t a b', positive=True, real=True)
r1 = sp.Matrix([-a * sp.sin(t), b * sp.cos(t), 0])
r2 = sp.Matrix([-a * sp.cos(t), -b * sp.sin(t), 0])
curvature = sp.simplify(sp.sqrt((r1.cross(r2)).dot(r1.cross(r2))) / sp.sqrt(r1.dot(r1))**3)
k_max = sp.simplify(curvature.subs(t, 0))
k_min = sp.simplify(curvature.subs(t, sp.pi / 2))
sp.Matrix([curvature, k_max, k_min])
```

### 1-4 答案

$r(t)=(a\cos t, b\sin t, 0)$

$r'(t)=(-a\sin t, b\cos t, 0), \quad r''(t)=(-a\cos t, -b\sin t, 0)$

$r'(t) \times r''(t)=(0,0,ab)$

$k(t)=\frac{|r'(t) \times r''(t)|}{|r'(t)|^3}=\frac{ab}{(a^2\sin^2 t+b^2\cos^2 t)^{3/2}}=\frac{ab}{[(a^2-b^2)\sin^2 t+b^2]^{3/2}}$

$k_{\max}=\frac{a}{b^2}, \quad k_{\min}=\frac{b}{a^2}$

### 2 题目

讨论当实数 $\alpha$ 取何值时，下列极限存在：$\lim\limits_{(x,y) \to (0,0)} f^{\alpha-1}(x,y)\sin\frac{1}{x^2+y^2}$，其中 $f(x,y)=\max\{|x|,|y|\}$。若极限存在，求其值。

```python
import sympy as sp
n = sp.symbols('n', positive=True, integer=True)
x_n = 1 / sp.sqrt(sp.pi / 2 + 2 * sp.pi * n)
phase = sp.simplify(sp.sin(1 / x_n**2))
case_alpha_2 = sp.limit(x_n * phase, n, sp.oo)
case_alpha_1 = sp.simplify(phase)
case_alpha_0 = sp.limit(phase / x_n, n, sp.oo)
sp.Matrix([phase, case_alpha_2, case_alpha_1, case_alpha_0])
```

### 2 答案

当 $\alpha>1$ 时，利用 $|\sin \frac{1}{x^2+y^2}| \le 1$ 以及 $f(x,y)=\max\{|x|,|y|\} \to 0$，可得

$|f^{\alpha-1}(x,y)\sin \frac{1}{x^2+y^2}| \le f^{\alpha-1}(x,y) \to 0$

因此极限存在且等于 $0$。

当 $\alpha \le 1$ 时，取 $x_n=\frac{1}{\sqrt{\pi/2+2n\pi}}, y_n=0$，则 $\sin \frac{1}{x_n^2}=1$，从而 $f^{\alpha-1}(x_n,0)\sin \frac{1}{x_n^2}=x_n^{\alpha-1}$。

若 $\alpha=1$，上式恒为 $1$；若 $\alpha<1$，上式发散到 $+\infty$。因此 $\alpha \le 1$ 时极限不存在。

综上，极限存在当且仅当 $\alpha>1$，且此时极限值为 $0$。

设 $f(x)$ 是区间 $[a,b]$ 上的有界函数，$D(f)$ 是 $f$ 的所有间断点构成的集合，则 $f(x)$ 于 $[a,b]$ 上 Riemann 可积的充分必要条件是：

### 3-1 答案

答案：D。

### 3-2 题目

设函数 $f(x)=(x-\frac{1}{2})^2 D(x)$，其中 $D(x)$ 为 Dirichlet 函数，即当 $x$ 为有理数时取值为 $1$，当 $x$ 为无理数时取值为 $0$。问 $f(x)$ 于区间 $[0,1]$ 上是否可积？给出理由。

```python
import sympy as sp
n = sp.symbols('n', positive=True, integer=True)
x0 = sp.Rational(1, 4)
t_n = x0 + 1 / n
rational_values = sp.simplify((t_n - sp.Rational(1, 2))**2)
irrational_values = sp.Integer(0)
sp.Matrix([sp.limit(rational_values, n, sp.oo), sp.limit(irrational_values, n, sp.oo)])
```

### 3-2 答案

$f$ 在 $[0,1]$ 上不可积。

对任意 $x_0 \in [0, \frac{1}{2}) \cup (\frac{1}{2},1]$，可以分别取趋于 $x_0$ 的无理数列与有理数列。沿无理数列时函数值恒为 $0$；沿有理数列时函数值趋于 $(x_0-\frac{1}{2})^2>0$。

因此 $f$ 在上述每一点都不连续，故 $[0, \frac{1}{2}) \cup (\frac{1}{2},1] \subseteq D(f)$。该集合不是零测集，所以 $D(f)$ 不是零测集。

由 Riemann 可积判别准则，$f$ 在 $[0,1]$ 上不可积。

### 4 题目

设 $C_i \subseteq \mathbb{R}^n$，$i=1,2,\cdots,m$，均为紧致集，且 $m$ 为有限正整数。证明这些点集的并集 $C=\bigcup_{i=1}^m C_i$ 为紧致集。

### 4 答案

任取 $C$ 的一个开覆盖。由于 $C_i \subseteq C$，这个开覆盖同时也是每个 $C_i$ 的开覆盖。

因为 $C_i$ 紧致，所以对每个 $i$ 都能从中选出有限个开集覆盖 $C_i$。再把这 $m$ 个有限子覆盖并在一起，仍然是有限个开集，并且覆盖整个 $C=\bigcup_{i=1}^m C_i$。

因此 $C$ 的任一开覆盖都有有限子覆盖，所以 $C$ 是紧致集。

$\operatorname{QED}$

### 5 题目

计算由平面曲线 $\frac{x^2}{a^2}-\frac{y^2}{b^2}=1$ 在区间 $a \le x \le 2a$ 上对应弧段绕 $x$ 轴旋转一周所得旋转体的体积与旋转面面积。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x, a, b = sp.symbols('x a b', positive=True, real=True)
f = b * sp.sqrt(x**2 / a**2 - 1)
volume = sp.simplify(sp.pi * sp.integrate(f**2, (x, a, 2 * a)))

k = a**2 + b**2
surface_integrand = b / a**2 * sp.sqrt(k * x**2 - a**4)
primitive = (
    sp.Rational(1, 2) * x * sp.sqrt(k * x**2 - a**4)
    - a**4 / (2 * sp.sqrt(k)) * sp.log(sp.sqrt(k) * x + sp.sqrt(k * x**2 - a**4))
)
surface = sp.simplify(
    2 * sp.pi * b / a**2 * (primitive.subs(x, 2 * a) - primitive.subs(x, a))
)

display(Math(rf"V = {sp.latex(volume)}"))
display(Math(rf"f(x)\\sqrt{{1+f'(x)^2}} = {sp.latex(surface_integrand)}"))
display(Math(rf"S = {sp.latex(surface)}"))

a_val = 1.0
b_val = 0.7
x_vals = np.linspace(a_val, 2 * a_val, 240)
profile = b_val * np.sqrt(x_vals**2 / a_val**2 - 1)

fig = plt.figure(figsize=(11, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x_vals, profile, label='upper branch')
ax1.plot(x_vals, -profile, label='lower branch')
ax1.fill_between(x_vals, 0, profile, color='tab:blue', alpha=0.2)
ax1.set_title('Hyperbola arc on $[a, 2a]$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True, alpha=0.3)
ax1.legend()

theta_vals = np.linspace(0, 2 * np.pi, 90)
X, Theta = np.meshgrid(x_vals, theta_vals, indexing='ij')
R = profile[:, None]
Y = R * np.cos(Theta)
Z = R * np.sin(Theta)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True, alpha=0.85)
ax2.set_title('Surface of revolution')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

plt.tight_layout()
plt.show()
```

### 5 答案

设 $f(x)=b\sqrt{\frac{x^2}{a^2}-1}$，则 $x \in [a,2a]$ 时有 $f'(x)=\frac{b}{a}\cdot \frac{x}{\sqrt{x^2-a^2}}$。

$$
\begin{aligned}
V &= \pi \int_a^{2a} f^2(x) \, dx = b^2 \pi \int_a^{2a} \left(\frac{x^2}{a^2}-1\right) dx = \frac{4}{3}ab^2\pi,
\end{aligned}
$$

$$
\begin{aligned}
S &= 2\pi \int_a^{2a} f(x)\sqrt{1+f'(x)^2} \, dx \\
  &= \frac{2\pi b}{a^2} \int_a^{2a} \sqrt{(a^2+b^2)x^2-a^4} \, dx \\
  &= 2\pi b\sqrt{3a^2+4b^2}-\pi b^2+\frac{\pi a^2 b}{\sqrt{a^2+b^2}} \ln \frac{\sqrt{a^2+b^2}+b}{2\sqrt{a^2+b^2}+\sqrt{3a^2+4b^2}}.
\end{aligned}
$$

### 6 题目

椭球面 $\frac{x^2}{a^2}+\frac{y^2}{b^2}+\frac{z^2}{c^2}=1$ 过点 $(1,1,2)$，且该点处切平面为 $4x+2y+3z=12$，求 $a,b,c$ 的值。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

A, B, C, k = sp.symbols('A B C k', positive=True, real=True)
solution = sp.solve(
    [
        sp.Eq(A, 2 * k),
        sp.Eq(B, k),
        sp.Eq(4 * C, 3 * k),
        sp.Eq(A + B + 4 * C, 1),
    ],
    [A, B, C, k],
    dict=True,
)[0]
a_sol = sp.simplify(sp.sqrt(1 / solution[A]))
b_sol = sp.simplify(sp.sqrt(1 / solution[B]))
c_sol = sp.simplify(sp.sqrt(1 / solution[C]))
display(Math(rf"a = {sp.latex(a_sol)},\quad b = {sp.latex(b_sol)},\quad c = {sp.latex(c_sol)}"))

a_val = float(a_sol)
b_val = float(b_sol)
c_val = float(c_sol)

u_vals = np.linspace(0, np.pi, 90)
v_vals = np.linspace(0, 2 * np.pi, 180)
U, V = np.meshgrid(u_vals, v_vals, indexing='ij')
X = a_val * np.sin(U) * np.cos(V)
Y = b_val * np.sin(U) * np.sin(V)
Z = c_val * np.cos(U)

x_plane = np.linspace(-0.3, 2.2, 50)
y_plane = np.linspace(-0.3, 2.4, 50)
X_plane, Y_plane = np.meshgrid(x_plane, y_plane, indexing='ij')
Z_plane = (12 - 4 * X_plane - 2 * Y_plane) / 3

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='Blues', alpha=0.38, linewidth=0)
ax.plot_surface(X_plane, Y_plane, Z_plane, color='tab:orange', alpha=0.4, linewidth=0)
ax.scatter([1], [1], [2], color='black', s=35)
ax.text(1.05, 1.05, 2.05, '(1, 1, 2)')
ax.set_title('Ellipsoid and tangent plane at $(1,1,2)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_box_aspect((a_val, b_val, c_val))
plt.tight_layout()
plt.show()
```

### 6 答案

设 $F(x,y,z)=\frac{x^2}{a^2}+\frac{y^2}{b^2}+\frac{z^2}{c^2}-1$。则椭球面在 $(1,1,2)$ 处的法向量为 $\left(\frac{2}{a^2}, \frac{2}{b^2}, \frac{4}{c^2}\right)$。

而切平面 $4x+2y+3z=12$ 的法向量为 $(4,2,3)$，故存在常数 $k$ 使得

$\left(\frac{2}{a^2}, \frac{2}{b^2}, \frac{4}{c^2}\right)=k(4,2,3)$

即 $\frac{1}{a^2}=2k, \frac{1}{b^2}=k, \frac{4}{c^2}=3k$。再由点 $(1,1,2)$ 在椭球面上，得 $\frac{1}{a^2}+\frac{1}{b^2}+\frac{4}{c^2}=1$。

代入得到 $k=\frac{1}{6}$，从而 $a=\sqrt{3}, \quad b=\sqrt{6}, \quad c=2\sqrt{2}$。

设映射 $f(x,y,z)=\binom{x+y+z}{xyz}$，

$$
g(u,v)=\begin{pmatrix}
a\sin u\cos v \\
a\sin u\sin v \\
a\cos u
\end{pmatrix}
$$

求复合映射 $f \circ g$ 的 Jacobi 矩阵。

```python
import sympy as sp
u, v, a = sp.symbols('u v a', positive=True, real=True)
x, y, z = sp.symbols('x y z', real=True)
f = sp.Matrix([x + y + z, x * y * z])
g = sp.Matrix([a * sp.sin(u) * sp.cos(v), a * sp.sin(u) * sp.sin(v), a * sp.cos(u)])
fog = f.subs({x: g[0], y: g[1], z: g[2]})
sp.simplify(fog.jacobian([u, v]))
```

### 7 答案

$f \circ g$ 的 Jacobi 矩阵为

$$
J(f \circ g)=\begin{pmatrix}
a\cos u(\sin v+\cos v)-a\sin u & a\sin u(\cos v-\sin v) \\
a^3\sin u\sin v\cos v(2\cos^2 u-\sin^2 u) & a^3\sin^2 u\cos u(\cos^2 v-\sin^2 v)
\end{pmatrix}.
$$

### 8-1 题目

设二元函数

$$
f(x,y)=\begin{cases}\frac{x^2y}{x^2+y^2}, & x^2+y^2 \ne 0 \\
0, & x^2+y^2=0
\end{cases}
$$

证明 $f(x,y)$ 于 $(0,0)$ 处沿任何方向的方向导数皆存在。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

t, theta = sp.symbols('t theta', real=True)
directional = sp.simplify(
    (((t * sp.cos(theta))**2 * (t * sp.sin(theta))) / ((t * sp.cos(theta))**2 + (t * sp.sin(theta))**2)) / t
)
display(Math(rf"\frac{{\partial f}}{{\partial u}}(0,0) = {sp.latex(directional)}"))

theta_vals = np.linspace(0, 2 * np.pi, 400)
directional_func = sp.lambdify(theta, directional, 'numpy')
directional_vals = directional_func(theta_vals)

x_vals = np.linspace(-1.0, 1.0, 220)
y_vals = np.linspace(-1.0, 1.0, 220)
X, Y = np.meshgrid(x_vals, y_vals, indexing='ij')
R2 = X**2 + Y**2
Z = np.divide(X**2 * Y, R2, out=np.zeros_like(X), where=R2 > 1e-12)

fig = plt.figure(figsize=(11, 4))
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='coolwarm', linewidth=0, antialiased=True, alpha=0.88)
ax1.set_title('Surface of $f(x,y)$ near $(0,0)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(theta_vals, directional_vals, color='tab:green')
ax2.set_title('Directional derivative versus direction')
ax2.set_xlabel(r'$\theta$')
ax2.set_ylabel(r'$\partial f / \partial u$')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 8-1 答案

设方向向量 $u=(\cos \theta, \sin \theta)$。则

$$
\frac{\partial f}{\partial u}(0,0)=\lim_{t \to 0} \frac{f(t\cos \theta, t\sin \theta)-f(0,0)}{t}=\cos^2 \theta \sin \theta.
$$

因此 $(0,0)$ 处沿任何方向的方向导数都存在。

### 8-2 题目

证明 $f(x,y)$ 于 $(0,0)$ 处不可微。

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = sp.symbols('x', positive=True)
remainder_y0 = sp.Integer(0)
remainder_yx = sp.simplify((x**3 / (2 * x**2)) / sp.sqrt(2 * x**2))
limit_y0 = sp.limit(remainder_y0, x, 0)
limit_yx = sp.limit(remainder_yx, x, 0)
display(Math(rf"\lim_{{x\to 0^+}} \frac{{f(x,0)}}{{\sqrt{{x^2}}}} = {sp.latex(limit_y0)}"))
display(Math(rf"\lim_{{x\to 0^+}} \frac{{f(x,x)}}{{\sqrt{{x^2+x^2}}}} = {sp.latex(limit_yx)}"))

x_vals = np.linspace(0.01, 0.6, 250)
ratio_y0 = np.zeros_like(x_vals)
ratio_yx = np.full_like(x_vals, float(sp.N(limit_yx)))

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(x_vals, np.zeros_like(x_vals), label=r'$y=0$')
axes[0].plot(x_vals, x_vals, label=r'$y=x$')
axes[0].scatter([0], [0], color='black', s=20)
axes[0].set_title('Two approach paths to $(0,0)$')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].axis('equal')
axes[0].grid(True, alpha=0.3)
axes[0].legend()

axes[1].plot(x_vals, ratio_y0, label=r'$\frac{f(x,0)}{\sqrt{x^2}}$')
axes[1].plot(x_vals, ratio_yx, label=r'$\frac{f(x,x)}{\sqrt{x^2+x^2}}$')
axes[1].set_title('Normalized quotient along the two paths')
axes[1].set_xlabel('x')
axes[1].set_ylabel(r'$f(x,y)/\sqrt{x^2+y^2}$')
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```

### 8-2 答案

先注意到 $\frac{\partial f}{\partial x}(0,0)=\frac{\partial f}{\partial y}(0,0)=0$。若 $f$ 在 $(0,0)$ 处可微，则应有

$$
\lim_{(x,y) \to (0,0)} \frac{f(x,y)}{\sqrt{x^2+y^2}}=0.
$$

但沿路径 $y=0$ 时，上式恒为 $0$；沿路径 $y=x$ 时，

$$
\frac{f(x,x)}{\sqrt{x^2+x^2}}=\frac{x^3/(2x^2)}{\sqrt{2}x}=\frac{\sqrt{2}}{4}.
$$

两条路径得到的极限不同，因此该极限不存在，故 $f$ 在 $(0,0)$ 处不可微。
