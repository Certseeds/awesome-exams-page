### 1-3 题目

设 $u, v$ 是由方程组 $\left\{\begin{array}{l}x u-y v=1 \\ y u+x v=0\end{array}\right.$ 所确定的 $x, y$ 的可微函数，求 $\frac{\partial u}{\partial x}, \frac{\partial u}{\partial y}, \frac{\partial v}{\partial x}, \frac{\partial v}{\partial y}$（设 $x^{2}+y^{2} \neq 0$）。

```python
import sympy as sp
from IPython.display import display, Math

x, y, u, v = sp.symbols("x y u v", real=True)
A = sp.Matrix([[x, -y], [y, x]])
B = sp.Matrix([[u, -v], [v, u]])
J = sp.simplify(-A.inv() * B)

display(Math(
    r"\left(\begin{array}{ll}\frac{\partial u}{\partial x} & \frac{\partial u}{\partial y}\\\frac{\partial v}{\partial x} & \frac{\partial v}{\partial y}\end{array}\right)="
    + sp.latex(J)
))
display(Math(r"\frac{\partial u}{\partial x}=" + sp.latex(sp.simplify(J[0, 0]))))
display(Math(r"\frac{\partial u}{\partial y}=" + sp.latex(sp.simplify(J[0, 1]))))
display(Math(r"\frac{\partial v}{\partial x}=" + sp.latex(sp.simplify(J[1, 0]))))
display(Math(r"\frac{\partial v}{\partial y}=" + sp.latex(sp.simplify(J[1, 1]))))
```

### 1-3 答案

设 $F_1(x, y, u, v)=x u-y v-1$，$F_2(x, y, u, v)=y u+x v$，则

$$
\left(\begin{array}{ll}\frac{\partial u}{\partial x} & \frac{\partial u}{\partial y} \\ \frac{\partial v}{\partial x} & \frac{\partial v}{\partial y}\end{array}\right)=-\left(\begin{array}{cc}\frac{\partial F_{1}}{\partial u} & \frac{\partial F_{1}}{\partial v} \\ \frac{\partial F_{2}}{\partial u} & \frac{\partial F_{2}}{\partial v}\end{array}\right)^{-1}\left(\begin{array}{ll}\frac{\partial F_{1}}{\partial x} & \frac{\partial F_{1}}{\partial y} \\ \frac{\partial F_{2}}{\partial x} & \frac{\partial F_{2}}{\partial y}\end{array}\right)
$$

$$
=-\left(\begin{array}{cc}x & -y \\ y & x\end{array}\right)^{-1}\left(\begin{array}{cc}u & -v \\ v & u\end{array}\right)
$$

$$
=-\frac{1}{x^{2}+y^{2}}\left(\begin{array}{ll}x u+y v & -x v+y u \\ x v-y u & x u+y v\end{array}\right)
$$

因此：

$$
\frac{\partial u}{\partial x}=-\frac{x u+y v}{x^{2}+y^{2}}, \quad \frac{\partial u}{\partial y}=\frac{x v-y u}{x^{2}+y^{2}}
$$

$$
\frac{\partial v}{\partial x}=\frac{-x v+y u}{x^{2}+y^{2}}, \quad \frac{\partial v}{\partial y}=-\frac{x u+y v}{x^{2}+y^{2}}
$$

### 1-4 题目

计算曲线积分 $\int_{L} \mathrm{e}^{\sqrt{x^{2}+y^{2}}} \mathrm{~d} s$，其中 $L$ 为圆周 $x^{2}+y^{2}=a^{2}$ 位于第一象限的部分。

```python
import sympy as sp
t, a = sp.symbols("t a", positive=True, real=True)
x_t = a * sp.cos(t)
y_t = a * sp.sin(t)
arc = sp.integrate(sp.exp(sp.sqrt(x_t**2 + y_t**2)) * sp.sqrt(sp.diff(x_t, t)**2 + sp.diff(y_t, t)**2), (t, 0, sp.pi / 2))
sp.simplify(arc)
```

### 1-4 答案

$$
\int_{L} e^{\sqrt{x^{2}+y^{2}}} \mathrm{d} s=\int_{0}^{\frac{\pi}{2}} a e^{a} \mathrm{d} t=\frac{\pi}{2} a e^{a}
$$

## 二 (12 分)

设 $f(x, y)=\frac{x y}{3 x^{2}+2 y^{2}}$（$(x, y) \neq(0,0)$）；$f(0,0)=0$。

### 2-1 题目

$f(x, y)$ 于 $(0,0)$ 处是否连续？给出理由。

```python
import sympy as sp
t = sp.symbols("t", real=True, nonzero=True)
f_axis = sp.simplify((t * 0) / (3 * t**2 + 2 * 0**2))
f_diag = sp.simplify((t * t) / (3 * t**2 + 2 * t**2))
sp.Matrix([f_axis, f_diag])
```

```python
import numpy as np
import matplotlib.pyplot as plt

grid = np.linspace(-1.0, 1.0, 220)
X, Y = np.meshgrid(grid, grid)
denominator = 3 * X**2 + 2 * Y**2
F = np.divide(X * Y, denominator, out=np.full_like(X, np.nan), where=denominator > 1e-4)

fig = plt.figure(figsize=(6.2, 5.0))
ax = fig.add_subplot(111, projection="3d")
surface = ax.plot_surface(X, Y, F, cmap="coolwarm", linewidth=0, antialiased=False, alpha=0.9)

t = np.linspace(-1.0, 1.0, 240)
ax.plot(t, np.zeros_like(t), np.zeros_like(t), color="black", linewidth=2, label="y=0")
ax.plot(t, t, 0.2 * np.ones_like(t), color="tab:green", linewidth=2, label="y=x")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title(r"$f(x,y)=\frac{xy}{3x^2+2y^2}$ near $(0,0)$")
fig.colorbar(surface, ax=ax, shrink=0.65, pad=0.08)
ax.legend()
plt.show()
```

### 2-1 答案

取 $S_{i}=(\frac{1}{i}, 0) \neq(0,0)$，$t_{i}=(\frac{1}{i}, \frac{1}{i}) \neq(0,0)$，则 $\lim _{i \rightarrow \infty} S_{i}=\lim _{i \rightarrow \infty} t_{i}=(0,0)$，而 $f(S_{i})=0$，$f(t_{i})=\frac{1}{5}$。

因此 $\lim _{(x, y) \rightarrow(0,0)} f(x, y)$ 不存在，所以 $f(x, y)$ 于 $(0,0)$ 处不连续。

### 2-2 题目

$\frac{\partial f}{\partial x}(0,0)$ 和 $\frac{\partial f}{\partial y}(0,0)$ 是否存在？如果存在，求出它们的值。

```python
import sympy as sp
t = sp.symbols("t", real=True)
quotient_x = sp.simplify((t * 0) / (3 * t**2 + 2 * 0**2))
quotient_y = sp.simplify((0 * t) / (3 * 0**2 + 2 * t**2))
sp.Matrix([sp.limit(quotient_x / t, t, 0), sp.limit(quotient_y / t, t, 0)])
```

### 2-2 答案

$$
\frac{\partial f}{\partial x}(0,0)=\lim _{t \rightarrow 0} \frac{f(t, 0)}{t}=0
$$

$$
\frac{\partial f}{\partial y}(0,0)=\lim _{t \rightarrow 0} \frac{f(0, t)}{t}=0
$$

### 4 题目

计算 $I=\iint_{D}(x^{2} y^{2}+x y) \mathrm{d} x \mathrm{~d} y$，其中 $D=\{(x, y):|x|+|y| \leq 1\}$。

```python
import sympy as sp
x, y = sp.symbols("x y", real=True, nonnegative=True)
result = 4 * sp.integrate(sp.integrate(x**2 * y**2, (y, 0, 1 - x)), (x, 0, 1))
sp.simplify(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_grid = np.linspace(0.0, 1.0, 220)
y_grid = np.linspace(0.0, 1.0, 220)
X, Y = np.meshgrid(x_grid, y_grid)
Z = X**2 * Y**2 + X * Y
Z = np.where(X + Y <= 1.0, Z, np.nan)

fig = plt.figure(figsize=(6.0, 5.0))
ax = fig.add_subplot(111, projection="3d")
surface = ax.plot_surface(X, Y, Z, cmap="viridis", linewidth=0, antialiased=False, alpha=0.92)

boundary_x = np.linspace(0.0, 1.0, 240)
boundary_y = 1.0 - boundary_x
boundary_z = boundary_x**2 * boundary_y**2 + boundary_x * boundary_y
ax.plot(boundary_x, boundary_y, boundary_z, color="black", linewidth=2)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel(r"$x^2 y^2 + xy$")
ax.set_title(r"$x^2 y^2 + xy$ on $x+y\leq 1$, $x,y\geq 0$")
fig.colorbar(surface, ax=ax, shrink=0.65, pad=0.08)
plt.show()
```

### 4 答案

由对称性知 $\iint_{D} x y \mathrm{d} x \mathrm{d} y=0$，$\iint_{D} x^{2} y^{2} \mathrm{d} x \mathrm{d} y=4 \iint_{\substack{x+y \leq 1 \\ x \geq 0, y \geq 0}} x^{2} y^{2} \mathrm{d} x \mathrm{d} y$。

因此：

$$
I=4 \iint_{\substack{x+y \leq 1 \\ x \geq 0, y \geq 0}} x^{2} y^{2} \mathrm{d} x \mathrm{d} y
$$

$$
=4 \int_{0}^{1} x^{2} \mathrm{d} x \int_{0}^{1-x} y^{2} \mathrm{d} y=\frac{4}{3} \int_{0}^{1} x^{2}(1-x)^{3} \mathrm{d} x=\frac{1}{45}
$$

## 五 (10 分)

### 5 题目

设 $f(x, y)$ 是 $D=\{(x, y): x^{2}+y^{2} \leq r^{2}\}$ 上的非负连续函数。证明：若 $A=\sqrt{a^{2}+b^{2}}>r$，则有

$$
\frac{m}{A+r} \pi r^{2} \leq \iint_{D} \frac{f(x, y)}{\sqrt{(x-a)^{2}+(y-b)^{2}}} \mathrm{~d} x \mathrm{~d} y \leq \frac{M}{A-r} \pi r^{2}
$$

其中 $M=\max \{f(x, y):(x, y) \in D\}$，$m=\min \{f(x, y):(x, y) \in D\}$。

```python
import numpy as np
import matplotlib.pyplot as plt

r = 2.0
a, b = 3.4, 1.2
theta = np.linspace(0, 2 * np.pi, 400)
fig, ax = plt.subplots(figsize=(5, 5))
ax.fill(r * np.cos(theta), r * np.sin(theta), alpha=0.15, color="tab:blue")
ax.plot(r * np.cos(theta), r * np.sin(theta), color="tab:blue", label=r"$D$")
ax.scatter([a], [b], color="tab:red", label=r"$P=(a,b)$")
for angle in (0.4, 2.2):
    x_pt = r * np.cos(angle)
    y_pt = r * np.sin(angle)
    ax.plot([a, x_pt], [b, y_pt], alpha=0.7, color="tab:gray")
ax.set_aspect("equal", adjustable="box")
ax.grid(alpha=0.3)
ax.legend()
plt.show()
```

### 5 答案

证明：

设 $P$ 的坐标为 $(a, b)$。则 $A=\sqrt{a^{2}+b^{2}}$ 为 $P$ 到原点的距离，$\sqrt{(x-a)^{2}+(y-b)^{2}}$ 为 $P$ 到 $D$ 中某一点的距离。由 $A>r$ 知 $P \notin D$，那么：

$$
A-r \leq \sqrt{(x-a)^{2}+(y-b)^{2}} \leq A+r
$$

因此：

$$
\frac{m}{A+r} \leq \frac{f(x, y)}{\sqrt{(x-a)^{2}+(y-b)^{2}}} \leq \frac{M}{A-r}
$$

对积分区域 $D$ 积分得：

$$
\iint_D \frac{m}{A+r} \mathrm{d} x \mathrm{d} y \leq \iint_D \frac{f(x, y)}{\sqrt{(x-a)^{2}+(y-b)^{2}}} \mathrm{d} x \mathrm{d} y \leq \iint_D \frac{M}{A-r} \mathrm{d} x \mathrm{d} y
$$

即：

$$
\frac{m}{A+r} \pi r^{2} \leq \iint_{D} \frac{f(x, y)}{\sqrt{(x-a)^{2}+(y-b)^{2}}} \mathrm{d} x \mathrm{d} y \leq \frac{M}{A-r} \pi r^{2}
$$

## 六 (10 分)

### 6 题目

计算 $I=\iint_{\Sigma} x^{2} \mathrm{~d} y \mathrm{~d} z+y^{2} \mathrm{~d} z \mathrm{~d} x+z^{2} \mathrm{~d} x \mathrm{~d} y$，其中 $\Sigma$ 是由曲面 $z=\sqrt{x^{2}+y^{2}}$ 和平面 $z=1$ 围成的三维有界区域的表面外侧。

```python
import sympy as sp
r, theta, z = sp.symbols("r theta z", real=True, nonnegative=True)
integrand = 2 * z * r
result = sp.integrate(sp.integrate(sp.integrate(integrand, (z, r, 1)), (r, 0, 1)), (theta, 0, 2 * sp.pi))
sp.simplify(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0.0, 2 * np.pi, 220)
radius = np.linspace(0.0, 1.0, 140)
Theta, R = np.meshgrid(theta, radius)
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
Z_cone = R
Z_plane = np.ones_like(R)

fig = plt.figure(figsize=(6.2, 5.2))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z_cone, cmap="Blues", linewidth=0, alpha=0.78)
ax.plot_surface(X, Y, Z_plane, cmap="Oranges", linewidth=0, alpha=0.42)
ax.plot(np.cos(theta), np.sin(theta), np.ones_like(theta), color="black", linewidth=1.6)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title(r"Region bounded by $z=\sqrt{x^2+y^2}$ and $z=1$")
ax.set_box_aspect((1, 1, 0.8))
plt.show()
```

### 6 答案

设上述区域为 $\Omega$，由 Gauss 公式及对称性知：

$$
I = \iint_{\Sigma} x^{2} \mathrm{d} y \mathrm{d} z + y^{2} \mathrm{d} z \mathrm{d} x + z^{2} \mathrm{d} x \mathrm{d} y = 2 \iiint_{\Omega}(x+y+z) \mathrm{d} x \mathrm{d} y \mathrm{d} z
$$

$$
= 2 \iiint_{\Omega} z \mathrm{d} x \mathrm{d} y \mathrm{d} z = 2 \iint_{x^2+y^2 \leq 1} \mathrm{d} x \mathrm{d} y \int_{\sqrt{x^{2}+y^{2}}}^{1} z \mathrm{d} z
$$

$$
= \iint_{x^{2}+y^{2} \leq 1}(1-(x^{2}+y^{2})) \mathrm{d} x \mathrm{d} y = \int_{0}^{2 \pi} \mathrm{d}\theta \int_{0}^{1}(1-r^{2}) r \mathrm{d} r = \frac{\pi}{2}
$$

### 7 题目

计算曲线积分 $\int_{\Gamma} \frac{y \mathrm{~d} x-x \mathrm{~d} y}{x^{2}+y^{2}}$，其中 $\Gamma$ 为平面上的闭曲线 $x^{2}+4 y^{2}-2 \sqrt{3} x-1=0$，方向为逆时针方向。

```python
import numpy as np
import sympy as sp

t = sp.symbols("t", real=True)
x_t = sp.sqrt(3) + 2 * sp.cos(t)
y_t = sp.sin(t)
integrand_expr = (y_t * sp.diff(x_t, t) - x_t * sp.diff(y_t, t)) / (x_t**2 + y_t**2)
integrand = sp.lambdify(t, integrand_expr, "numpy")

grid = np.linspace(0, 2 * np.pi, 4001)
x_vals = np.sqrt(3) + 2 * np.cos(grid)
y_vals = np.sin(grid)
angles = np.unwrap(np.arctan2(y_vals, x_vals))
winding_number = int(round((angles[-1] - angles[0]) / (2 * np.pi)))
numerical_value = np.trapezoid(integrand(grid), grid)

{
    "numerical_integral": numerical_value,
    "winding_number": winding_number,
    "exact_value": -2 * sp.pi * winding_number
}
```

### 7 答案

作一个以原点为中心、以充分小的 $\varepsilon>0$ 为半径的圆周 $C_{\varepsilon}$，正向取顺时针，设 $C_{\varepsilon}$ 与 $\Gamma$ 之间的区域为 $D$。由 Green 公式：

$$
\int_{\Gamma} \frac{y \mathrm{d} x - x \mathrm{d} y}{x^{2}+y^{2}} = -\int_{C_{\varepsilon}} \frac{y \mathrm{d} x - x \mathrm{d} y}{x^{2}+y^{2}} = -\frac{1}{\varepsilon^{2}} \int_{C_{\varepsilon}} y \mathrm{d} x - x \mathrm{d} y = -\frac{2\pi\varepsilon^{2}}{\varepsilon^{2}} = -2\pi
$$

### 8 题目

用 Lagrange 乘数法求平面曲线 $a x^{2}+2 b x y+c y^{2}=1$ 上的点到原点的最远距离和最近距离，其中 $a, b, c$ 皆为正数，且 $a c-b^{2}>0$。

```python
import sympy as sp
from IPython.display import display, Math

a, b, c = sp.symbols("a b c", positive=True, real=True)
Delta = sp.sqrt((a - c)**2 + 4 * b**2)
r_max_sq = sp.radsimp(2 / (a + c - Delta))
r_min_sq = sp.radsimp(2 / (a + c + Delta))
r_max = sp.sqrt(r_max_sq)
r_min = sp.sqrt(r_min_sq)

display(Math(r"x^{2}+y^{2}_{\max}=" + sp.latex(r_max_sq)))
display(Math(r"x^{2}+y^{2}_{\min}=" + sp.latex(r_min_sq)))
display(Math(r"r_{\max}=" + sp.latex(r_max)))
display(Math(r"r_{\min}=" + sp.latex(r_min)))
```

### 8 答案

设 $f(x, y)=x^{2}+y^{2}$，

$F(x, y)=f(x, y)+\lambda(a x^{2}+2 b x y+c y^{2}-1)$，

$=(a \lambda+1) x^{2}+2 b \lambda x y+(c \lambda+1) y^{2}-\lambda$。

令：

$$
\frac{\partial F}{\partial x}=2[(a \lambda+1) x+b \lambda y]=0 \quad \quad \quad (1)
$$

$$
\frac{\partial F}{\partial y}=2[b \lambda x+(c \lambda+1) y]=0 \quad \quad \quad (2)
$$

约束条件：

$$
a x^{2}+2 b x y+c y^{2}=1
\Leftrightarrow
\left(\begin{array}{ll}x & y\end{array}\right)\left(\begin{array}{ll}a & b \\ b & c\end{array}\right)\left(\begin{array}{l}x \\ y\end{array}\right)=1
$$

由于 $\left(\begin{array}{ll}a & b \\ b & c\end{array}\right)$ 是正定阵，故特征值均大于 0，因此该平面曲线是椭圆，为有界闭集。由于连续函数在有界闭集上必然取得最大值和最小值，并且 $x=0, y=0$ 不满足 $a x^{2}+2 b x y+c y^{2}=1$，因此上述方程组有非零解，故：

$$
\left|\begin{array}{ll}a \lambda+1 & b \lambda \\ b \lambda & c \lambda+1\end{array}\right|=(a \lambda+1)(c \lambda+1)-b^{2} \lambda^{2}=(a c-b^{2}) \lambda^{2}+(a+c) \lambda+1=0
$$

因此：

$$
\lambda=\frac{-(a+c) \pm \sqrt{(a-c)^{2}+4 b^{2}}}{2(a c-b^{2})}
$$

由 (1) 可得 $\frac{y}{x}=-\frac{a \lambda+1}{b \lambda}$。

由 (2) 可得 $\frac{x}{y}=-\frac{c \lambda+1}{b \lambda}$。

因此：

$$
x^{2}+y^{2}=\frac{x^{2}+y^{2}}{a x^{2}+2 b x y+c y^{2}}=\frac{\frac{x}{y}+\frac{y}{x}}{a \frac{x}{y}+2 b+c \frac{y}{x}}
$$

$$
=\frac{-\frac{c \lambda+1}{b \lambda}-\frac{a \lambda+1}{b \lambda}}{-a \cdot \frac{c \lambda+1}{b \lambda}+2 b-c \cdot \frac{a \lambda+1}{b \lambda}}=\frac{(a+c) \lambda+2}{2(a c-b^{2}) \lambda+(a+c)}
$$

$$
=\lambda \cdot \frac{(a+c) \lambda+2}{2(a c-b^{2}) \lambda^{2}+(a+c) \lambda}
$$

$$
=\lambda \cdot \frac{(a+c) \lambda+2}{-2[(a+c) \lambda+1]+(a+c) \lambda}
$$

$$
=\lambda \cdot \frac{(a+c) \lambda+2}{-(a+c) \lambda-2}=-\lambda
$$

因此 $x^{2}+y^{2}$ 的最大值为 $\frac{a+c+\sqrt{(a-c)^{2}+4 b^{2}}}{2(a c-b^{2})}$，最小值为 $\frac{a+c-\sqrt{(a-c)^{2}+4 b^{2}}}{2(a c-b^{2})}$。

所以平面曲线 $a x^{2}+2 b x y+c y^{2}=1$ 上的点到原点的最远距离为：

$$
\sqrt{\frac{a+c+\sqrt{(a-c)^{2}+4 b^{2}}}{2(a c-b^{2})}}
$$

最近距离为：

$$
\sqrt{\frac{a+c-\sqrt{(a-c)^{2}+4 b^{2}}}{2(a c-b^{2})}}
$$
