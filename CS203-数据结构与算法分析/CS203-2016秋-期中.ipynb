### 1-1 题目

已知入栈序列依次为 $a,b,c,d$，请选择所有不可能出现的出栈序列。

```python
from itertools import permutations

push_seq = ['a', 'b', 'c', 'd']
invalid = []
choices = {
    'A': ['a', 'd', 'b', 'c'],
    'B': ['a', 'd', 'c', 'b'],
    'C': ['b', 'a', 'c', 'd'],
    'D': ['b', 'd', 'a', 'c'],
    'E': ['b', 'd', 'c', 'a'],
}
for label, perm in choices.items():
    stack = []
    i = 0
    ok = True
    for target in perm:
        while (not stack or stack[-1] != target) and i < len(push_seq):
            stack.append(push_seq[i])
            i += 1
        if stack and stack[-1] == target:
            stack.pop()
        else:
            ok = False
            break
    if not ok:
        invalid.append(label)
print(invalid)
```

### 1-1 答案

答案：A、D。

### 1-2 题目

请写出中缀表达式 $3+((1+2)*2-2)*3$ 所对应的后缀表达式。

```python
expr = '3+((1+2)*2-2)*3'
precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
stack, output = [], []
for ch in expr:
    if ch.isdigit():
        output.append(ch)
    elif ch == '(':
        stack.append(ch)
    elif ch == ')':
        while stack[-1] != '(':
            output.append(stack.pop())
        stack.pop()
    else:
        while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[ch]:
            output.append(stack.pop())
        stack.append(ch)
while stack:
    output.append(stack.pop())
print(''.join(output))
```

答案：`312+2*2-3*+`。

### 1-3 题目

请写出后缀表达式 `3451-*+3-` 的计算结果。

```python
def eval_postfix(expr):
    stack = []
    for ch in expr:
        if ch.isdigit():
            stack.append(int(ch))
        else:
            b = stack.pop()
            a = stack.pop()
            stack.append(eval(f'{a}{ch}{b}'))
    return stack[-1]
print(eval_postfix('3451-*+3-'))
```

### 1-4 题目

设栈 S 和队列 Q 的初始状态为空，元素 $e_1$ 到 $e_6$ 依次通过栈；一个元素出栈后即进入队列 Q。若 6 个元素出队的序列为 $e_2,e_4,e_3,e_6,e_5,e_1$，则栈 S 的容量至少应该是多少？

```python
push_order = ['e1', 'e2', 'e3', 'e4', 'e5', 'e6']
pop_order = ['e2', 'e4', 'e3', 'e6', 'e5', 'e1']
stack = []
next_push = 0
max_size = 0
trace = []

for target in pop_order:
    while (not stack or stack[-1] != target) and next_push < len(push_order):
        stack.append(push_order[next_push])
        max_size = max(max_size, len(stack))
        trace.append(f'push {push_order[next_push]} -> {stack[:]}')
        next_push += 1
    popped = stack.pop()
    trace.append(f'pop  {popped} -> {stack[:]}')

for step in trace:
    print(step)
print('minimum capacity =', max_size)
```

### 1-4 答案

答案：`3`。

### 1-8-2 题目

请根据 1-8-1中的 $\operatorname{next}(\mathrm{n})$ 数组, 请仿照下图画出匹配过程, 即用图示表示出每一轮匹配 T 在 S 中的位置, 并写出下标


```python
def build_next(pattern):
    n = len(pattern)
    nxt = [0] * n
    nxt[0] = -1
    i, j = 0, -1
    while i < n - 1:
        if j == -1 or pattern[i] == pattern[j]:
            i += 1; j += 1
            nxt[i] = j
        else:
            j = nxt[j]
    return nxt

S = 'aababcbcabcababc'
T = 'abcaba'
nxt = build_next(T)

# KMP 匹配并记录每轮对齐位置和匹配长度
i, j = 0, 0
rounds = []  # (align_pos, matched_count)
while i < len(S) and j < len(T):
    if j == -1 or S[i] == T[j]:
        i += 1; j += 1
    else:
        rounds.append((i - j, j))  # 对齐位置, 已匹配数
        j = nxt[j]
if j == len(T):
    rounds.append((i - j, len(T)))

# 打印匹配过程
sep = '|'
header = f'{"":>5}{sep}' + sep.join(S) + sep
print(header)
for k, (align, mlen) in enumerate(rounds):
    row = [' '] * len(S)
    show = mlen if mlen == len(T) else mlen + 1  # 显示匹配部分 + 失配字符
    for c in range(min(show, len(T))):
        if align + c < len(S):
            row[align + c] = T[c]
    tag = f'Loop{k + 1}'
    note = f'start={align}' + (', matched!' if mlen == len(T) else '')
    print(f'{tag:>5}{sep}' + sep.join(row) + f'{sep} {note}')
```

### 1-8-2 答案

答案：首次匹配成功的位置是下标 `8`。若按题目图示书写匹配过程，可依次给出模式串在主串上的对齐位置，直到在下标 `8` 处完全匹配。

### 1-8-3 题目

请写出简单匹配算法(朴素匹配算法)的时间复杂度 $\underline{\quad\quad\quad}$ (简单匹配算法请见下图). 根据 2)中匹配规律, 写出 KMP 算法时间复杂度 $\underline{\quad\quad\quad}$

||a|a|b|a|b|c|b|c|a|b|c|a|b|a|b|c|
|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
|Loop1|a|b|c|a|b|a||||||||||0|
|Loop2||a|b|c|a|b|a||||||||||
|Loop3|||a|b|c|a|b|a|||||||||
|...|||||||||||||||||
|Loop9|||||||||a|b|c|a|b|a|||

```python
def naive_match(text, pattern):
    comparisons = 0
    for start in range(len(text) - len(pattern) + 1):
        matched = True
        for offset, char in enumerate(pattern):
            comparisons += 1
            if text[start + offset] != char:
                matched = False
                break
        if matched:
            return start, comparisons
    return -1, comparisons

def build_next(pattern):
    nxt = [-1] * len(pattern)
    j = -1
    for i in range(1, len(pattern)):
        while j >= 0 and pattern[j + 1] != pattern[i]:
            j = nxt[j]
        if pattern[j + 1] == pattern[i]:
            j += 1
        nxt[i] = j
    return nxt

def kmp_match(text, pattern):
    nxt = build_next(pattern)
    j = -1
    comparisons = 0
    for i, ch in enumerate(text):
        while j >= 0 and pattern[j + 1] != ch:
            comparisons += 1
            j = nxt[j]
        comparisons += 1
        if pattern[j + 1] == ch:
            j += 1
            if j == len(pattern) - 1:
                return i - j, comparisons, nxt
    return -1, comparisons, nxt

text = 'aaaaabaaaaabaaaaab'
pattern = 'aaab'
print('naive  -> index, comparisons =', naive_match(text, pattern))
index, comparisons, nxt = kmp_match(text, pattern)
print('kmp    -> index, comparisons =', (index, comparisons))
print('next array =', nxt)
print('time complexity: naive O(MN), KMP O(M+N)')
```

### 1-8-3 答案

答案：朴素匹配算法为 $O(MN)$，KMP 算法为 $O(M+N)$。

### 2-1 题目

对一个整数连续做“除 2 之后再减 1”的运算，连续做 `n` 次之后结果为 `1`。求初始值，并分别用递归和非递归方式实现。

```python
def f_recursive(n):
    if n == 0:
        return 1
    return 2 * (f_recursive(n - 1) + 1)

def f_iterative(n):
    ans = 1
    for _ in range(n):
        ans = 2 * (ans + 1)
    return ans
print(f_recursive(3), f_iterative(3))
```

### 2-1 答案

逆向思考：若一次操作后得到 `y`，则操作前的数满足 `x = 2(y+1)`。因此从结果 `1` 开始反推 `n` 次即可。

+ 递归：`f(n) = 2 * (f(n-1) + 1)`，`f(0)=1`
+ 非递归：从 `ans = 1` 开始循环执行 `ans = 2 * (ans + 1)`

### 2-2 题目

建立两个单链表，长度未知。设计算法找出两个单链表中第一个公共节点，并尽可能降低时间复杂度。

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

def length(head):
    count = 0
    while head is not None:
        count += 1
        head = head.next
    return count

def first_common_node(head1, head2):
    n1, n2 = length(head1), length(head2)
    if n1 < n2:
        head1, head2 = head2, head1
        n1, n2 = n2, n1
    for _ in range(n1 - n2):
        head1 = head1.next
    while head1 is not None and head2 is not None:
        if head1 is head2:
            return head1
        head1 = head1.next
        head2 = head2.next
    return None

common = Node('C1', Node('C2'))
head1 = Node('A1', Node('A2', common))
head2 = Node('B1', common)
node = first_common_node(head1, head2)
print('first common node =', node.value if node else None)
```

### 2-2 答案

先分别求出两条链表长度，让较长链表先走 `|m-n|` 步；随后两个指针同步前进，第一次相遇的结点就是第一个公共结点。时间复杂度 $O(m+n)$。

### 2-3 题目

利用已经定义好的 `double` 型栈结构，实现根据已知后缀表达式求出运算结果。

```python
def eval_postfix(expr):
    stack = []
    for ch in expr:
        if ch.isdigit():
            stack.append(float(ch))
        else:
            b = stack.pop()
            a = stack.pop()
            stack.append(eval(f'{a}{ch}{b}'))
    return stack[-1]
print(eval_postfix('23+4*'))
```

### 2-3 答案

算法思路：从左到右扫描后缀表达式。若读到操作数则压栈；若读到运算符，则弹出两个栈顶元素进行计算，再把结果压栈。最后栈顶元素即为表达式值。

### 2-4 题目

给出链式队列的类定义，并写出相应的入队和出队方法。

```python
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

class LinkedQueue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, value):
        node = Node(value)
        if self.rear is None:
            self.front = self.rear = node
        else:
            self.rear.next = node
            self.rear = node

    def dequeue(self):
        if self.front is None:
            return None
        value = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return value

    def to_list(self):
        values = []
        current = self.front
        while current is not None:
            values.append(current.data)
            current = current.next
        return values

queue = LinkedQueue()
for value in [10, 20, 30]:
    queue.enqueue(value)
    print(f'enqueue {value} -> {queue.to_list()}')
print('dequeue ->', queue.dequeue(), queue.to_list())
queue.enqueue(40)
print('enqueue 40 ->', queue.to_list())
```

### 2-4 答案

链式队列需要维护 `front` 和 `rear` 两个指针。入队时在 `rear` 后插入新结点并更新 `rear`；出队时删除 `front` 所指向的结点，若删除后队空，还需要把 `rear` 置空。

### 2-5 题目

设计算法判断字符串是否含有回文子串，例如 `analyze` 为包含，`alone` 为不包含。

```python
def has_palindrome_substring(s):
    n = len(s)
    for center in range(n):
        l = r = center
        while l >= 0 and r < n and s[l] == s[r]:
            if r - l + 1 >= 2:
                return True
            l -= 1
            r += 1
        l, r = center, center + 1
        while l >= 0 and r < n and s[l] == s[r]:
            if r - l + 1 >= 2:
                return True
            l -= 1
            r += 1
    return False
print(has_palindrome_substring('analyze'), has_palindrome_substring('alone'))
```

### 2-5 答案

可采用中心扩展法。对每个位置分别检查奇数长度回文和偶数长度回文，只要存在长度至少为 2 的回文子串就返回真。时间复杂度 $O(n^2)$，空间复杂度 $O(1)$。
