### 1-1 题目

表达式 $a*(b+c)-d$ 的后缀表达式为 $\underline{\quad\quad\quad}$.

````
expr = 'a*(b+c)-d'
precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
stack, output = [], []
for ch in expr:
    if ch.isalnum():
        output.append(ch)
    elif ch == '(':
        stack.append(ch)
    elif ch == ')':
        while stack[-1] != '(':
            output.append(stack.pop())
        stack.pop()
    else:
        while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[ch]:
            output.append(stack.pop())
        stack.append(ch)
while stack:
    output.append(stack.pop())
print(''.join(output))
````

### 1-1 答案

答案：`abc+*d-`

````
from itertools import permutations

chars = ['A', 'B', 'C', 'D', 'E']
results = []
for perm in permutations(chars):
    stack = []
    i = 0
    ok = True
    for target in perm:
        while (not stack or stack[-1] != target) and i < len(chars):
            stack.append(chars[i])
            i += 1
        if stack and stack[-1] == target:
            stack.pop()
        else:
            ok = False
            break
    if ok and perm[0] == 'C' and perm[1] == 'D':
        results.append(''.join(perm))
print(results)
````

### 1-10 题目

已知某通信电文只有 6 种字符 $\{a,b,c,d,e,f\}$ 构成, 每个字符出现的次数分别为 $\{23,5,14,8,25,7\}$, 请给出各字符的 Huffman 编码：$\underline{\quad\quad\quad}$.

````
import heapq

freqs = {'a': 23, 'b': 5, 'c': 14, 'd': 8, 'e': 25, 'f': 7}
heap = []
for i, (char, weight) in enumerate(sorted(freqs.items())):
    heapq.heappush(heap, (weight, i, char))
count = len(heap)
while len(heap) > 1:
    w1, _, left = heapq.heappop(heap)
    w2, _, right = heapq.heappop(heap)
    heapq.heappush(heap, (w1 + w2, count, (left, right)))
    count += 1
codes = {}
def build_codes(node, code=''):
    if isinstance(node, str):
        codes[node] = code or '0'
        return
    left, right = node
    build_codes(left, code + '0')
    build_codes(right, code + '1')
build_codes(heap[0][2])
print(', '.join(f'{char}:{codes[char]}' for char in sorted(codes)))
````

### 1-10 答案

答案：一组可行的 Huffman 编码为 `a:10, b:0110, c:00, d:010, e:11, f:0111`。

### 2-6 题目

对某有向无环图(DAG)进行拓扑排序, 可以得到的不同拓扑序列的个数是

```log
node a, b,c,e,d
edge:
a to b
a to c
a to e
b to c
c to d
e to d
```

+ A 4
+ B 3
+ C 2
+ D 1

````
nodes = ['a', 'b', 'c', 'd', 'e']
edges = [('a', 'b'), ('a', 'c'), ('a', 'e'), ('b', 'c'), ('c', 'd'), ('e', 'd')]

def topo_count(remaining):
    if not remaining:
        return 1
    indeg = {x: 0 for x in remaining}
    for u, v in edges:
        if u in remaining and v in remaining:
            indeg[v] += 1
    available = [x for x in remaining if indeg[x] == 0]
    return sum(topo_count([y for y in remaining if y != x]) for x in available)

print(topo_count(nodes))
````

答案：B。不同拓扑序列共有 `3` 个。

````
expr = 'a/b+(c*d-e*f)/g'
precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
stack = []
for ch in expr:
    if ch.isalnum():
        if ch == 'f':
            break
    elif ch == '(':
        stack.append(ch)
    elif ch == ')':
        while stack[-1] != '(':
            stack.pop()
        stack.pop()
    else:
        while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[ch]:
            stack.pop()
        stack.append(ch)
print(''.join(stack))
````

试证明，在具有 $n(n\ge 1)$ 个结点的 $m$ 叉树中，有 $n(m-1)+1$ 个指针域为空。

````
def empty_pointer_count(n, m):
    return n * m - (n - 1)

for n, m in [(1, 3), (5, 2), (7, 4)]:
    print(n, m, empty_pointer_count(n, m))
````

### 3 答案

每个结点都有 $m$ 个指针域，因此全树一共有 $mn$ 个指针域。

树中共有 $n-1$ 条边，而每一条边恰好对应一个非空指针域，所以空指针域个数为

$$
mn-(n-1)=n(m-1)+1
$$

证毕。

### 4-1 题目

给定关键字序列 $T=[13,25,16,29,8,28,4,9,20,12,19]$，使用快速排序并选择第一个记录作为 `pivotkey`，写出第一趟排序结束时的序列。

````
T = [13, 25, 16, 29, 8, 28, 4, 9, 20, 12, 19]

def partition_first_pass(arr):
    arr = arr[:]
    low, high = 0, len(arr) - 1
    pivot = arr[low]
    while low < high:
        while low < high and arr[high] >= pivot:
            high -= 1
        arr[low] = arr[high]
        while low < high and arr[low] <= pivot:
            low += 1
        arr[high] = arr[low]
    arr[low] = pivot
    return arr

print(partition_first_pass(T))
````

答案：第一趟结束后的序列为 `[12, 9, 4, 8, 13, 28, 29, 16, 20, 25, 19]`。

### 4-2 题目

对同一序列使用希尔排序，增量序列取 $\{4,2,1\}$，写出每轮排序结束时的序列。

````
T = [13, 25, 16, 29, 8, 28, 4, 9, 20, 12, 19]

def shell_sort_rounds(arr, gaps):
    arr = arr[:]
    rounds = []
    for gap in gaps:
        for i in range(gap, len(arr)):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        rounds.append((gap, arr[:]))
    return rounds

print(shell_sort_rounds(T, [4, 2, 1]))
````

### 4-2 答案

+ 增量 $1$：`[4, 8, 9, 12, 13, 16, 19, 20, 25, 28, 29]`

### 4-3 题目

对同一序列使用基数排序（最低位优先），写出每轮排序结束时的序列。

````
T = [13, 25, 16, 29, 8, 28, 4, 9, 20, 12, 19]

def radix_sort_rounds(arr):
    arr = arr[:]
    exp = 1
    rounds = []
    while max(arr) // exp > 0:
        buckets = [[] for _ in range(10)]
        for num in arr:
            buckets[(num // exp) % 10].append(num)
        arr = [num for bucket in buckets for num in bucket]
        rounds.append((exp, arr[:]))
        exp *= 10
    return rounds

print(radix_sort_rounds(T))
````

### 4-3 答案

答案：

+ 个位排序后：`[20, 12, 13, 4, 25, 16, 8, 28, 29, 9, 19]`
+ 十位排序后：`[4, 8, 9, 12, 13, 16, 19, 20, 25, 28, 29]`

### 5 题目

已知数组 $A[1\ldots n]$ 中的元素均为正整数，且既包含奇数也包含偶数。请设计一个算法将其调整为左右两个部分，左边为奇数，右边为偶数，要求时间复杂度 $O(n)$，空间复杂度 $O(1)$。

````
def partition_odd_even(arr):
    arr = arr[:]
    left, right = 0, len(arr) - 1
    while left < right:
        while left < right and arr[left] % 2 == 1:
            left += 1
        while left < right and arr[right] % 2 == 0:
            right -= 1
        if left < right:
            arr[left], arr[right] = arr[right], arr[left]
    return arr

sample = [2, 7, 4, 9, 6, 11, 8, 5]
print(partition_odd_even(sample))
````

### 5 答案

基本思想：设置左右两个指针，左指针向右寻找偶数，右指针向左寻找奇数；当二者都停下且尚未相遇时交换两个元素。这样每次交换都会把一个奇数放到左边、一个偶数放到右边。

复杂度：时间复杂度 $O(n)$，空间复杂度 $O(1)$。

### 6 题目

已知带头结点的单链表，请在不改变链表节点分配的前提下，设计高效算法查找链表倒数第 $k$ 个位置上的节点（$k$ 为正整数）。成功返回该节点 `data`，否则返回 `0`。

````
class ListNode:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

def kth_from_end(head, k):
    fast = head.next
    slow = head.next
    for _ in range(k - 1):
        if fast is None:
            return 0
        fast = fast.next
    if fast is None:
        return 0
    while fast.next is not None:
        fast = fast.next
        slow = slow.next
    return slow.data if slow is not None else 0

head = ListNode(None)
head.next = ListNode(10, ListNode(20, ListNode(30, ListNode(40, ListNode(50)))))
print([kth_from_end(head, k) for k in [1, 2, 5, 6]])
````

### 6 答案

基本思想：使用 `fast` 和 `slow` 两个指针。先让 `fast` 比 `slow` 领先 $k-1$ 步；之后同时前进，当 `fast` 到达尾结点时，`slow` 恰好位于倒数第 $k$ 个结点。

复杂度：时间复杂度 $O(n)$，空间复杂度 $O(1)$。

### 7 题目

判别序列 `[14,63,8,35,28,46,13,82,88,21]` 是否为最小堆。若不是，请调整为最小堆并画出调整过程。

````
heap = [14, 63, 8, 35, 28, 46, 13, 82, 88, 21]
steps = []

def sift_down(arr, i, n):
    while True:
        left = 2 * i + 1
        right = 2 * i + 2
        smallest = i
        if left < n and arr[left] < arr[smallest]:
            smallest = left
        if right < n and arr[right] < arr[smallest]:
            smallest = right
        if smallest == i:
            return
        arr[i], arr[smallest] = arr[smallest], arr[i]
        steps.append(arr[:])
        i = smallest

for i in range(len(heap) // 2 - 1, -1, -1):
    sift_down(heap, i, len(heap))
print(steps)
print(heap)
````

### 7 答案

该序列不是最小堆，因为根结点 `14` 大于其右孩子 `8`。

调整过程：

1. `[14, 63, 8, 35, 21, 46, 13, 82, 88, 28]`
2. `[14, 21, 8, 35, 63, 46, 13, 82, 88, 28]`
3. `[14, 21, 8, 35, 28, 46, 13, 82, 88, 63]`
4. `[8, 21, 14, 35, 28, 46, 13, 82, 88, 63]`
5. `[8, 21, 13, 35, 28, 46, 14, 82, 88, 63]`

最终最小堆：`[8, 21, 13, 35, 28, 46, 14, 82, 88, 63]`。

### 8 题目

给出带权有向图，使用 Dijkstra 算法求从顶点 $V_0$ 到其它顶点的最短距离和最短路径。

````
edges = {
    'v0': [('v2', 10), ('v4', 30), ('v5', 100)],
    'v1': [('v2', 5)],
    'v2': [('v3', 50)],
    'v3': [('v5', 10)],
    'v4': [('v3', 20), ('v5', 50)],
    'v5': []
}
vertices = ['v0', 'v1', 'v2', 'v3', 'v4', 'v5']

def dijkstra(start):
    dist = {v: float('inf') for v in vertices}
    prev = {v: None for v in vertices}
    used = set()
    dist[start] = 0
    for _ in vertices:
        cur = min((v for v in vertices if v not in used), key=lambda v: dist[v])
        used.add(cur)
        for nxt, weight in edges[cur]:
            if dist[cur] + weight < dist[nxt]:
                dist[nxt] = dist[cur] + weight
                prev[nxt] = cur
    return dist, prev

dist, prev = dijkstra('v0')
print(dist)
print(prev)
````

### 8 答案

最短距离与路径如下：

+ $V_0$：距离 $0$，路径 `V0`
+ $V_1$：不可达
+ $V_2$：距离 $10$，路径 `V0 -> V2`
+ $V_3$：距离 $50$，路径 `V0 -> V4 -> V3`
+ $V_4$：距离 $30$，路径 `V0 -> V4`
+ $V_5$：距离 $60$，路径 `V0 -> V4 -> V3 -> V5`

### 9 题目

对于二叉搜索树(Binary Search Tree)，请设计算法将其按结点 `data` 值递增顺序转换为双向链表，并返回链表头结点。要求在原结点结构上做修改，不能新建新的链表结点。

````
class Node:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def tree_to_doubly_list(root):
    head = None
    prev = None
    def inorder(node):
        nonlocal head, prev
        if node is None:
            return
        inorder(node.left)
        if prev is None:
            head = node
        else:
            prev.right = node
            node.left = prev
        prev = node
        inorder(node.right)
    inorder(root)
    return head

root = Node(40, Node(20, Node(15), Node(24)), Node(50, Node(45), Node(60)))
head = tree_to_doubly_list(root)
result = []
while head is not None:
    result.append(head.data)
    head = head.right
print(result)
````

### 9 答案

基本思想：对二叉搜索树做中序遍历。由于 BST 的中序序列天然递增，只需在遍历过程中把当前结点与前驱结点原地相连，即可得到双向链表。

对题目示例，转换结果为：`15 <-> 20 <-> 24 <-> 40 <-> 45 <-> 50 <-> 60`。

复杂度：时间复杂度 $O(n)$，递归栈空间复杂度 $O(h)$。
