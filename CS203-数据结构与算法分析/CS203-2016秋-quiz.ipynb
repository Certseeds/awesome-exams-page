## 一 二叉树的性质

### 1-1 题目

在二叉树中，以根节点为第 0 层，则第 1 层上有多少个结点？

````python
def max_nodes_at_level(level):
    return 2 ** level

for level in range(5):
    print(f'level {level}: at most {max_nodes_at_level(level)} nodes')
````

### 1-1 答案

答案：第 1 层最多有 `2` 个结点。

### 1-2 题目

根结点为第 0 层，深度为层数最大的叶结点层数，则深度为 `k` 的二叉树至多有多少个结点？

````python
def max_nodes(k):
    return 2 ** (k + 1) - 1
print(max_nodes(3))
````

### 1-2 答案

答案：$2^{k+1}-1$。

### 1-3 题目

一棵二叉树，若其叶子结点数为 $n_0$，度为 2 的结点数为 $n_2$，则 $n_0$ 和 $n_2$ 的关系是什么？

````python
tree = {
    'A': ('B', 'C'),
    'B': ('D', 'E'),
    'C': (None, None),
    'D': (None, None),
    'E': ('F', 'G'),
    'F': (None, None),
    'G': (None, None),
}

def degree_counts(tree):
    n0 = 0
    n2 = 0
    for left, right in tree.values():
        degree = int(left is not None) + int(right is not None)
        if degree == 0:
            n0 += 1
        if degree == 2:
            n2 += 1
    return n0, n2

n0, n2 = degree_counts(tree)
print('n0 =', n0)
print('n2 =', n2)
print('n0 == n2 + 1 ->', n0 == n2 + 1)
````

### 1-3 答案

答案：$n_0 = n_2 + 1$。

### 1-4 题目

根结点为第 0 层，深度为层数最大的叶结点层数，则有 `n` 个结点的完全二叉树的深度为多少？

````python
import math

def depth_complete_tree(n):
    return math.floor(math.log2(n))
print(depth_complete_tree(10))
````

### 1-4 答案

答案：$\lfloor \log_2 n \rfloor$。

### 2 题目

请写出二叉树的非递归中序遍历算法。

````python
def inorder(root):
    stack = []
    cur = root
    while cur is not None or stack:
        while cur is not None:
            stack.append(cur)
            cur = cur.left
        cur = stack.pop()
        print(cur.value)
        cur = cur.right
````

### 2 答案

可用栈模拟递归：不断沿左孩子入栈，到空后弹出访问，再转向右子树。时间复杂度 $O(n)$。

### 3 题目

请写出二叉树的层序遍历算法。

````python
from collections import deque

def level_order(root):
    if root is None:
        return
    q = deque([root])
    while q:
        cur = q.popleft()
        print(cur.value)
        if cur.left: q.append(cur.left)
        if cur.right: q.append(cur.right)
````

### 3 答案

使用队列即可：根结点先入队，之后反复出队并把左右孩子按顺序入队。

已知一棵树的先序遍历序列和中序遍历序列分别为 `ABDGHCEFI` 和 `GDHBAECIF`，请画出此二叉树。

````python
pre = 'ABDGHCEFI'
ino = 'GDHBAECIF'
class Node:
    def __init__(self, v, l=None, r=None):
        self.v = v
        self.l = l
        self.r = r

def build(pre, ino):
    if not pre:
        return None
    root = pre[0]
    idx = ino.index(root)
    return Node(root, build(pre[1:1+idx], ino[:idx]), build(pre[1+idx:], ino[idx+1:]))

def to_tuple(node):
    if not node:
        return None
    return (node.v, to_tuple(node.l), to_tuple(node.r))
print(to_tuple(build(pre, ino)))
````

### 4 答案

对应二叉树为：

```log

       / \
      B   C
     /   / \
    D   E   F
   / \     /
  G   H   I
```

### 5 题目

简述最小堆的定义，并画出向一个空堆中逐步插入 `12,70,8,65,24,56,10,92,86,33` 的调整过程。

````python
seq = [12,70,8,65,24,56,10,92,86,33]
heap = []
states = []
for x in seq:
    heap.append(x)
    i = len(heap) - 1
    while i > 0:
        p = (i - 1) // 2
        if heap[p] <= heap[i]:
            break
        heap[p], heap[i] = heap[i], heap[p]
        i = p
    states.append(heap[:])
for state in states:
    print(state)
````

### 5 答案

最小堆定义：每个结点的关键字都不大于其左右孩子结点的关键字。

逐步插入后的堆状态为：

1. `[12]`
2. `[12, 70]`
3. `[8, 70, 12]`
4. `[8, 65, 12, 70]`
5. `[8, 24, 12, 70, 65]`
6. `[8, 24, 12, 70, 65, 56]`
7. `[8, 24, 10, 70, 65, 56, 12]`
8. `[8, 24, 10, 70, 65, 56, 12, 92]`
9. `[8, 24, 10, 70, 65, 56, 12, 92, 86]`
10. `[8, 24, 10, 70, 33, 56, 12, 92, 86, 65]`

### 6 题目

若森林中有 `n` 个结点和 `b` 条边（$b<n$），则该森林中有多少棵树？

````python
nodes = list(range(1, 8))
print(nodes)
edges = [(1, 2), (2, 3), (4, 5), (6, 7)]

def forest_tree_count(node_count, edge_count):
    return node_count - edge_count

def component_count(nodes, edges):
    parent = {node: node for node in nodes}

    def find(node):
        while parent[node] != node:
            parent[node] = parent[parent[node]]
            node = parent[node]
        return node

    def union(a, b):
        root_a = find(a)
        root_b = find(b)
        if root_a != root_b:
            parent[root_b] = root_a

    for u, v in edges:
        union(u, v)
    return len({find(node) for node in nodes})

print('n - b =', forest_tree_count(len(nodes), len(edges)))
print('connected components in sample forest =', component_count(nodes, edges))
````

### 6 答案

答案：`n-b`。因为每棵树的边数都等于结点数减一。

### 7 题目

请简单描述二叉搜索树的定义，并写出将指定 `K` 值插入到二叉搜索树中的算法。

````python
def insert(root, k):
    if root is None:
        return Node(k)
    if k < root.v:
        root.l = insert(root.l, k)
    elif k > root.v:
        root.r = insert(root.r, k)
    return root
````

### 7 答案

二叉搜索树满足: 
1. 左子树所有结点都小于根结点
2. 右子树所有结点都大于根结点。

插入时从根开始比较，递归或迭代地进入左/右子树，直到遇到空位置插入。

### 8 题目

已知二叉树按二叉链表形式存储，设计算法求二叉树第 `k` 层（根结点为第 0 层）上结点的个数。

````python
def count_level(root, k):
    if root is None:
        return 0
    if k == 0:
        return 1
    return count_level(root.l, k - 1) + count_level(root.r, k - 1)
````

### 8 答案

递归求解：当 `k=0` 时返回 1；否则答案等于左子树第 `k-1` 层结点数与右子树第 `k-1` 层结点数之和。

## 九

### 9 题目

已知二叉树按二叉链表形式存储，设计递归算法求出给定二叉树的深度（只含根节点的树深度为 0）。

````python
def depth(root):
    if root is None:
        return -1
    return max(depth(root.l), depth(root.r)) + 1
````

### 9 答案

答案：递归地分别求左右子树深度，取较大者再加 1。空树深度记为 `-1`，则仅含根结点的树深度为 `0`。
