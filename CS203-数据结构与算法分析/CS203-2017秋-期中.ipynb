### 1-2 题目

What is the corresponding postfix expression of the infix expression $(3+4*2)*3-6/3$: $\underline{\quad\quad\quad}$.

```python
expr = '(3+4*2)*3-6/3'
precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
stack, output = [], []
for ch in expr:
    if ch.isdigit():
        output.append(ch)
    elif ch == '(':
        stack.append(ch)
    elif ch == ')':
        while stack[-1] != '(':
            output.append(stack.pop())
        stack.pop()
    else:
        while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[ch]:
            output.append(stack.pop())
        stack.append(ch)
while stack:
    output.append(stack.pop())
print(''.join(output))
```

答案：`342*+3*63/-`。

### 1-5 题目

Suppose there is a stack; when we push an element into the stack it prints `S`, and when we pop an element it prints `X`. Four integers are pushed in order `1,2,3,4`. If we want the pop order `3,4,2,1`, what is the corresponding sequence of operations?

```python
push_seq = [1, 2, 3, 4]
pop_order = [3, 4, 2, 1]
stack = []
ops = []
push_idx = 0

for target in pop_order:
    while (not stack or stack[-1] != target) and push_idx < len(push_seq):
        stack.append(push_seq[push_idx])
        ops.append('S')
        push_idx += 1
    stack.pop()
    ops.append('X')

print(''.join(ops))
```

### 1-6 题目

Suppose an empty stack receives pushes $e_1,e_2,e_3,e_4,e_5,e_6$ in that order. If the pop order is $e_4,e_3,e_2,e_6,e_5,e_1$ what is the minimum capacity of the stack?

```python
push_order = ['e1', 'e2', 'e3', 'e4', 'e5', 'e6']
pop_order = ['e4', 'e3', 'e2', 'e6', 'e5', 'e1']
stack = []
next_push = 0
max_size = 0
trace = []

for target in pop_order:
    while (not stack or stack[-1] != target) and next_push < len(push_order):
        stack.append(push_order[next_push])
        max_size = max(max_size, len(stack))
        trace.append(f'push {push_order[next_push]} -> {stack[:]}')
        next_push += 1
    popped = stack.pop()
    trace.append(f'pop  {popped} -> {stack[:]}')

for step in trace:
    print(step)
print('minimum capacity =', max_size)
```

### 1-6 答案

答案：`4`。

### II-1 题目

A is a linear table $[a_1,a_2,...,a_n]$ and uses a sequential storage structure. In the case of equal probability, what's the average number of elements needed to move when adding an element to A?

```python
def average_moves(n):
    moves = [n - position for position in range(n + 1)]
    return moves, sum(moves) / len(moves)

for n in [4, 5, 8]:
    moves, average = average_moves(n)
    print(f'n={n}, moves by insertion position = {moves}, average = {average}')
```

### II-1 答案

答案：平均移动次数为 `n/2`。若向长度为 `n` 的顺序表插入，插入位置等概率，则需要移动的元素个数在 `0` 到 `n` 之间均匀分布，其平均值为 `n/2`。

### II-2-1 题目

Give the `next` array of the pattern string `P = "abababc"`.

```python
def build_next(pattern):
    n = len(pattern)
    nxt = [0] * n
    nxt[0] = -1
    i, j = 0, -1
    while i < n - 1:
        if j == -1 or pattern[i] == pattern[j]:
            i += 1; j += 1
            nxt[i] = j
        else:
            j = nxt[j]
    return nxt

P = 'abababc'
nxt = build_next(P)
print('下标 :', list(range(len(P))))
print('P[j] :', list(P))
print('next :', nxt)
```

### II-3 题目

Write down a class of doubly linked list and describe its process of adding and deleting an element.

```python
class Node:
    def __init__(self, item, prev=None, next=None):
        self.item = item
        self.prev = prev
        self.next = next

class DoublyLinkedList:
    def __init__(self):
        self.first = None
        self.last = None

    def add_last(self, item):
        node = Node(item, self.last, None)
        if self.last is None:
            self.first = node
        else:
            self.last.next = node
        self.last = node
        return node

    def insert_after(self, node, item):
        new_node = Node(item, node, node.next)
        if node.next is None:
            self.last = new_node
        else:
            node.next.prev = new_node
        node.next = new_node
        return new_node

    def remove(self, node):
        if node.prev is None:
            self.first = node.next
        else:
            node.prev.next = node.next
        if node.next is None:
            self.last = node.prev
        else:
            node.next.prev = node.prev

    def to_list(self):
        values = []
        current = self.first
        while current is not None:
            values.append(current.item)
            current = current.next
        return values

dll = DoublyLinkedList()
first = dll.add_last('A')
second = dll.add_last('B')
dll.add_last('D')
dll.insert_after(second, 'C')
print('after insert_after(B, C) =', dll.to_list())
dll.remove(first)
print('after remove(A) =', dll.to_list())
dll.remove(dll.last)
print('after remove(last) =', dll.to_list())
```

### II-3 答案

答案：双向链表结点同时保存 `prev` 和 `next` 指针。插入结点时，要同时维护新结点、前驱结点和后继结点之间的双向引用；删除结点时，也要把前驱与后继重新连起来，并更新头尾指针。

### III-1 题目

Given a nonnegative integer (1<=n<=10), compute `n!` using recursion.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

### III-1 答案

答案：递归边界为 `n == 1` 时返回 `1`；递归式为 `n! = n * (n-1)!`。

### III-2 题目

Suppose we have two singly linked lists. Maybe they have some common nodes. Design an algorithm to find their first common node.

```python
class Node:
    def __init__(self, item, next=None):
        self.item = item
        self.next = next

def length(head):
    count = 0
    while head is not None:
        count += 1
        head = head.next
    return count

def first_common_node(head1, head2):
    n1, n2 = length(head1), length(head2)
    if n1 < n2:
        head1, head2 = head2, head1
        n1, n2 = n2, n1
    for _ in range(n1 - n2):
        head1 = head1.next
    while head1 is not None and head2 is not None:
        if head1 is head2:
            return head1
        head1 = head1.next
        head2 = head2.next
    return None

common = Node('X', Node('Y'))
head1 = Node('A', Node('B', common))
head2 = Node('C', common)
node = first_common_node(head1, head2)
print('first common node =', node.item if node else None)
```

### III-2 答案

答案：先分别计算两条链表长度，让较长链表先走 `|m-n|` 步，再让两个指针同步前进；第一次相遇的结点就是第一个公共结点。时间复杂度 $O(m+n)$，空间复杂度 $O(1)$。

### III-3 题目

Using the defined double stack structure, implement how to calculate the result of a postfix expression.

```python
def eval_postfix(expr):
    stack = []
    for ch in expr:
        if ch.isdigit():
            stack.append(float(ch))
        else:
            b = stack.pop()
            a = stack.pop()
            stack.append(eval(f'{a}{ch}{b}'))
    return stack[-1]
print(eval_postfix('342*+'))
```

### III-3 答案

答案：从左到右扫描后缀表达式。遇到操作数就压栈；遇到运算符就弹出两个栈顶元素进行计算，再把结果压回栈中。扫描结束后，栈顶元素就是最终结果。

### III-4 题目

Find the largest minimum distance when placing `m` houses among positions `x1...xn`, with required time complexity $O(n\log n)$.

```python
def can_place(xs, m, gap):
    count = 1
    last = xs[0]
    for x in xs[1:]:
        if x - last >= gap:
            count += 1
            last = x
    return count >= m

def largest_min_distance(xs, m):
    xs = sorted(xs)
    left, right = 0, xs[-1] - xs[0]
    ans = 0
    while left <= right:
        mid = (left + right) // 2
        if can_place(xs, m, mid):
            ans = mid
            left = mid + 1
        else:
            right = mid - 1
    return ans

print(largest_min_distance([2, 4, 8, 10, 14], 3))
```

### III-4 答案

答案：先对位置排序，再对最小间距做二分搜索。对每个候选间距 `mid`，从左到右贪心放置房屋，检查是否至少能放下 `m` 栋。若可行则继续增大间距，否则减小间距。总复杂度为 $O(n\log n)$。
